```python
y = df.loc[df.groupby('Name')['Total Meta Weight'].tail(1).index, 'Total Meta Weight']

x = df.groupby('Name').apply(lambda x: set(x['Total Meta Weight']).pop() - x['Rounded Col A'].sum())

x = dict(x)
```

```python
df['Ouput Round A'] = df['Rounded Col A'].round(2).groupby(df['Name']).apply(lambda x:x)
```

```python
val = []
for indx,row in df.iterrows():
    if indx in y.index:
        val.append(row['Ouput Round A'] + x[row['Name']])
    else:
        val.append(row['Ouput Round A'])
```

```python
df['Ouput Round A'] = val
```

```python
def get_data(drid):
    mimetype ='application/vnd.google-apps.folder'
    query = f"parents = '{drid}' and mimeType = '{mimetype}' and trashed = false"
    response = service.files().list(q=query, driveId=drid, corpora='drive',includeItemsFromAllDrives=True,supportsAllDrives=True).execute()
    files = response.get('files')
    return files
```

```python
def generate_newsheet(service,data_df, file_rename,client,shared_drive_id,output_folder_id):
    """
    Copy the downloaded GHX data file to google spread sheet for Blue Prism team to pick up for their process.
    """
    file_metadata_OP = {'name' : file_rename.split(".")[0]  ,'parents': [output_folder_id],'mimeType': 'application/vnd.google-apps.spreadsheet','driveId': shared_drive_id}
    response = service.files().create(body=file_metadata_OP, supportsAllDrives=True).execute()
    ws= client.open_by_key(response['id']).worksheet("Sheet1")
    gd.set_with_dataframe(ws, data_df)
    return
```

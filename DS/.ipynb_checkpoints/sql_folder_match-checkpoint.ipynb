```python
import pandas as pd
import numpy as np
df_sql = pd.read_csv('df_sql.csv')
df_folder = pd.read_csv('fodler1.csv')
```

```python
df_sql['Title'] = df_sql['table_name'].apply(
    lambda x: ' '.join([i.strip() for i in x.split('-')[2:]]) 
    if x.count('-')>=2 else x.split('-')[1]
    )

df_sql['Title'] = df_sql['Title'].apply(lambda x: x.strip())

df_folder['Title'] = df_folder['team'].apply(
                    lambda x: ' '.join(x.split('-')[1:]) if '-' in x else x.split('-')[0]
                    )
df_folder['Title'] = df_folder['Title'].apply(lambda x: x.strip())
```

```python
unique_title_sql = set(list(df_sql['Title']))
unique_title_folder = set(list(df_folder['Title']))

title_common  = unique_title_sql.intersection(unique_title_folder)

All_Non_Matching_Title = unique_title_sql.symmetric_difference(unique_title_folder)

df_sql['Title_common'] = df_sql['Title'].apply(
            lambda x : x if x in title_common else float('NaN')
            )


df_folder['Title_common'] = df_folder['Title'].apply(
                            lambda x : x if x in title_common else float('NaN')
                            )
```

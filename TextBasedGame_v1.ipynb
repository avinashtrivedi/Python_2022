```python
# store all the movements into dictionary
rooms = {
        'Main Hall' : { 'South' : 'Kitchen', 
                       'North': 'Bedroom-1', 
                       'East' : 'Living Room', 
                       'West' : 'Library' },
    
        'Bedroom-1' : { 'South' : 'Main Hall', 
                       'East' : 'Basement', 
                       'Item' : 'Sword' },
    
        'Basement' : {'West': 'Bedroom-1',
                      'Item': 'Gun'},
        'Library': {'East':'Main Hall',
                    'Item':'Ammunition'},
    
        'Kitchen':{'East':'Master Bedroom',
                   'North':'Main Hall',
                   'Item':'Knives'},
    
        'Master Bedroom' : { 'West' : 'Kitchen', 
                            'Item' : "Chemical's for radiation cure"},
    
        'Living Room' : { 'West' : 'Main Hall',
                         'North': 'Garage', 
                         'Item' : 'Radiation cure shot' },
        'Garage':{'South':'Living Hall',
                  'Item':'Radiated Ghoul Monster'},
        }

# initial room
Curr_state = 'Main Hall'

# store collected items 
inventory = []

# function to move to new room
def get_new_room(state, direction):
    new_state = state 
    for i in rooms:  
        if i == state:  
            if direction in rooms[i]: 
                new_state=rooms[i][direction] 

    return new_state  

# get room item
def get_item(state):
    try:
        return rooms[state]['Item'] 
    except:
        return None

# function to print menu
def show_instructions():
    print("Ghoul Text Adventure Game")
    print("Collect 6 items to win the game, or be eaten by the Ghoul.")
    print("Move commands: go South, go North, go East, go West")
    print("Add to Inventory: get 'item name'")
    
show_instructions() 

# loop to keep asking for new commands
while True:  
    print('You are in ', Curr_state)  
    print('Inventory:',inventory) 
    
    # get item name from a room
    item = get_item(Curr_state) 
    if item:
        print('You see a',item) 
        if  item=='Radiated Ghoul Monster': 
            print('NOM NOM...GAME OVER!')
            print('Thanks for playing the game. Hope you enjoyed it.')
            break
    
    # receive command from user
    direction = input('Enter your move: ')  
    
    # get the command i.e either get or go along with name
    if direction.find(" ") != -1:
        indx_space = direction.find(" ")
        command = direction[:indx_space]
        name = direction[indx_space+1:]
    else:
        print('Wrong command')
        continue
    
    # to goto other room
    if command == 'go':
        if name in ('East','West','North','South'):
            new_state = get_new_room(Curr_state, name)  
            if new_state == Curr_state:  
                print("Invalid direction!")
            else:
                Curr_state = new_state  
        else:
             print("Invalid direction!")
    
    # to get the item from a room
    elif command == 'get' and 'Item' in rooms[Curr_state]:
            
            if name == rooms[Curr_state]['Item']:
                if item in inventory:
                    print('Item has been already taken, Please go to another room!!')
                else:
                    inventory.append(item)
            else:
                print("Wrong Item name")  
    else:
        print('Wrong command')
        
    # when number of collected items are 6
    if len(inventory)==6:
            print('Congratulations! You have collected all items and defeated the Ghoul!') 
            print('Thanks for playing the game. Hope you enjoyed it.')
            break
```

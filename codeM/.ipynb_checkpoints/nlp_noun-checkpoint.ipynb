```python
import spacy
import pandas as pd
nlp = spacy.load('en_core_web_sm')
```

```python
# load the dataset
df = pd.read_csv('sentiment2.csv')

# get only first 500 rows
df = df[:500]
```

```python
df
```

```python
# create new column for Noun Words
# store those words as list elements
df['Noun_Word'] = df['title'].apply(lambda x: [token.text for token in nlp(x) if token.pos_ == 'NOUN'])
```

```python
df
```

```python
df.to_csv('new_file_csv_v1.csv')
df.to_excel('new_file_excel_v1.xlsx')
```

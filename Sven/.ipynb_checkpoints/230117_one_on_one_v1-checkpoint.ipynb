```python
# Zahlen auf 2 Stellen runden

df = pd.read_excel(r'work\ubersicht.xlsx',
                    sheet_name='Benutzer',
                    usecols=[1, 2 ,3, 5],
                    header=None,
                    names=['fil','name', 'persnr', 'stunden']
                   )

gruppe = 1000
monat = 10
jahr = 2022


nicht_filialeliste = ['88', '95', '98', '99']

# Zeile mit der ursprünlichen Beschriftung löschensamessystem_lohn.py
df = df.drop([0])

# Löscht löscht leere Zeilen
df = df[df['persnr'].notna()]
df = df[df['fil'].notna()]

# Kostenstelle aus Filialnummer ableiten
df['fil'] = df['fil'].astype(str)
df['kst'] = 'k' + df['fil'].str[-2:]

df['monat'] = monat
df['jahr'] = jahr


df['stunden'] = df['stunden'].astype(float).round(2)

# nur ausgewählte Filialen, alle ausser nicht_filialliste
df = df[~df['fil'].isin(nicht_filialeliste)]

empl = df.pivot_table(index=['persnr', 'name'],
               columns=['kst'],
               values=['stunden'],
               fill_value=0,
               aggfunc='sum',
               margins_name = "anw",
               margins=True)

empl.reset_index(inplace=True, level=['persnr'])
# empl.reset_index(inplace=True)
# print(empl)

df_neu = empl
df_neu['gruppe'] = gruppe
df_neu['monat'] = monat
df_neu['jahr'] = jahr
# print(df_neu)
```

```python
df
```

```python
empl
```

```python
df_neu.columns = [''.join(col) for col in df_neu.columns]

df_neu.rename(columns=lambda s: s.replace("stunden", ""), inplace=True)


df_neu.to_csv("work/test.csv",
              sep=";",
              encoding='cp1252')
```

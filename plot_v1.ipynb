```python
def f(x):
    if x<100:
        return "0-100"
    elif x>=100 or x<200:
        return "100-200"
    elif x>=200 or x<300:
        return "200-300"
    elif x>=300 or x<400:
        return "300-400"
    elif x>=400 or x<500:
        return "400-500"
    else:
        return "500-1000"
```

```python
df['Fare_range'] = df['Fare'].apply(lambda x: f(x))
```

```python
plt.hist(df['Fare_range'])
plt.show()
```

```python
def f(x):
    if x<10:
        return "Kid"
    elif x>=10 or x<25:
        return "Teen"
    elif x>=25 or x<50:
        return "Adult"
    else:
        return "Senior Citizen"
```

```python
df['Age_group'] = df['Age'].apply(lambda x: f(x))
```

```python
plt.hist(df['Age'],color='red',label='all passengers')
plt.hist(df[df['Sex']=='female']['Age'],color='green',label='female')
plt.legend()
plt.show()
```

```python
from math import pi
from math import exp
from scipy import stats

# https://mathinsight.org/probability_distribution_idea
def normal_pdf(x, mu, sigma):
    return 1.0 / (sigma * (2.0 * pi)**(1/2)) * exp(-1.0 * (x - mu)**2 / (2.0 * (sigma**2)))
```

```python
s = []
for i in x:
    s.append(normal_pdf(i, x.mean(), x.std()))
```

```python
sum(s)
```

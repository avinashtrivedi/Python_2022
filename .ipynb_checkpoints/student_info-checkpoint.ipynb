```python
print("Welcome to the student data entry application.")
n_grade = int(input("How many grades will be entered for each student?: "))
print("\n")
print(" ok! Let's get started...")
print("\n")
i = 1
all_student_info = []
while True:
    student_info = []
    print(f"Enter Data for Student # {i}:")
    i = i+1
    name = input("Student's Name: ")
    id_num = input("Student's ID number: ")
    student_info.append(name)
    student_info.append(id_num)
    
    grade_info = []
    for j in range(n_grade):
        grade = int(input(f"Enter grade #{j+1}: "))
        grade_info.append(grade)
        
    student_info.append(grade_info)
    all_student_info.append(student_info)
    choice = input("Should we add another student? (y/n): ")
    print('\n')
    
    if choice in ('n','N'):
        print('--- Input Completed ---\n')
        break
        
print("Here is the review of the student data...")
for i in range(len(all_student_info)):
    student = all_student_info[i]
    name = student[0]
    id_num = student[1]
    grade = student[2]
    
    print(f'Student #{i+1}:')
    print('\tName:',name)
    print('\tID:',id_num)
    print('\tGrades:',grade)

print('\n--- End of the Program ---')
```

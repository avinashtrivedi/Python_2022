```python
# Start writing code here...# Wandelt eine Apollon-Auswertung in Zahlen um, die von Hutner verwendet werden können
# SG/9.11.2022
import pandas as pd
import numpy as np

excluded = [95, 96, 97, 98, 99, 780]

# Daten einlesen und vorbereiten
sales = pd.read_csv(r'D:\OneDrive - NITT\Custom_Download\Umsaetze_neu.csv',
                        encoding='cp1252',
                        sep=';',
                        dayfirst=True,
                        parse_dates=['DATE'],
                        dtype={6: 'str'},
                        low_memory=False
                        )
sales['ART'] = sales['ART'].astype(str)
sales['WG'] = sales['ART'].str[0]
sales['WG'] = sales['WG'].astype(int)
sales = sales.drop(sales.columns[[8, 9]], axis=1)


values = ['Fournituren', 'Schuhe', 'Accessoires']

conditions = [
              (sales['WG'] == 0), 
              (sales['WG'] > 0) & (sales['WG'] < 9),
              (sales['WG'] == 9)
              ]

sales['section'] = np.select(conditions, values) 
```

# Method-1

```python
sales.groupby('FIL').agg(
             a_sum=('BON', 'sum'),
             a_mean=('MEN', 'mean'),
             b_mean=('WG', 'mean'),
             d_range=('WG', lambda x: x.max() - x.min())
)
```

# Method-2

```python
def f(x):
    d = {}
    d['a_sum'] = x['BON'].sum()
    d['a_max'] = x['MEN'].max()
    d['b_mean'] = x['WG'].mean()
    return pd.Series(d, index=['a_sum', 'a_max', 'b_mean'])

sales.groupby('FIL').apply(f)
```

```python
import h5py 
import numpy as np
import pandas as pd
import glob
from tqdm.notebook import tqdm
```

```python
timestamps=pd.read_csv('time.csv')
```

```python
file = h5py.File('PS1.mat')
values = file['test1'][()]
data=pd.DataFrame(values.T)
data['date_time']=timestamps
data.set_index('date_time', inplace=True)
data.columns=['occup','mean','max','min','std']
```

```python
data
```

```python
# data.dropna(inplace=True)
```

```python
data_copy = data.copy()
```

```python
data= data_copy.diff()
```

```python
def eligible(row_indx):
    return all([False if abs(val)>=5 else True for val in list(data.loc[row_indx])])
```

```python
# data = data [:10000]
```

```python
lst = []
for i in tqdm(data.index):
    lst.append(eligible(i))
```

```python
data.reset_index(inplace=True)
data['date_time'] = pd.to_datetime(data['date_time'])
```

```python
data['date'] = data['date_time'].dt.date
```

```python
all_date = pd.date_range(start=data.date_time.min().date(), end=data.date_time.max().date())
all_date = [date.date() for date in all_date]
```

```python
all_date
```

```python
new_lst = lst.copy()
for i,val in enumerate(lst):
    if val == True and i>0:
        new_lst[i-1] = True
```

```python
data['combine'] = new_lst
```

```python
new_lst = []
i = 0
last_status = True
for val in data['combine']:
    if val == True:
        if last_status == False:
            i = i + 1
            new_lst.append(i)
        else:
            new_lst.append(i)
            
        last_status = True
    else:
        i = i + 1
        new_lst.append(i)
        last_status = False
data['group_col'] = new_lst
```

```python
data_copy['group_col'] = data['group_col'].values
```

```python
data_copy.head(130)
```

```python
df = data_copy[:]
```

```python
df.reset_index(inplace=True)
df['date_time'] = pd.to_datetime(df['date_time'])
```

```python
# df.groupby('group_col').mean()

df_final = df.groupby('group_col').agg(
                        date=('date_time',lambda x:list(set(x.dt.date))[0]),
                        time_start=('date_time',lambda x: x.min()),
                        time_end=('date_time',lambda x: x.max()),
                        occup_avg=('occup', 'mean'),
                        mean_avg=('mean', 'mean'),
                        max_avg=('max', 'mean'),
                        min_avg=('min', 'mean'),
                        std_avg=('std','mean'),
                        total_count=('group_col','count')
)
```

```python
missing_line = sorted(list(set(all_date) - set(data['date'])))
```

```python
missing_line
```

```python
df_final['Status'] = 'Up'
```

```python
ddff = []
for line in tqdm(all_date):
    if line not in missing_line:
        small_df = df_final[df_final['date']==line]
        ddff.append(small_df)
    else:
        d = pd.DataFrame([[line]*3 + [float('NaN')]*6 + ['Down']],columns=list(df_final))
        ddff.append(d)
```

```python
df_full = pd.concat(ddff)
df_full.reset_index(drop=True,inplace=True)
```

```python
df_full.drop('date',axis=1,inplace=True)
```

```python
def days_hours_minutes(td):
    d,h,m =  td.days, td.seconds//3600, (td.seconds//60)%60
    if d==0 and h==0 and m==0:
        return f'{m} Minutes'
    if d==0 and h==0:
        return f'{m} Minutes'
    elif d==0:
        return f"{h} Hour, {m} Minutes"
    else:
        return f"{d} Day,{h} Hour,{m} Minutes"
```

```python
df_full['Duration'] = df_full['time_end'] - df_full['time_start']
```

```python
df_full['Duration']  = df_full['Duration'].map(days_hours_minutes)
```

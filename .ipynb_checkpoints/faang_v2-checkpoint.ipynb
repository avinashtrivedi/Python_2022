# Q3
- Build a crosstab with the earthquake data between the tsunami column and the magType column. Rather than showing the frequency count, show the maximum magnitude that was observed for each combination. Put the magnitude type along with the columns.

```python
pd.crosstab(earthquakes['tsunami'], earthquakes['magType'], aggfunc='max',values=earthquakes['mag'],
            rownames=['tsunami'], colnames=['magType'])
```

# Q4
- Calculate the rolling 30-day aggregations of the OHLC data by ticker for the FAANG data. Use the following aggregation,Mean of the opening price

Maximum of the high price

Minimum of the low price

Mean of the closing price

Sum of the volume traded

```python
import pandas as pd
from collections import OrderedDict

faang['date'] = pd.to_datetime(faang['date'])
```

```python
faang
```

```python
for tkr in set(faang['ticker']):
    roll = faang[faang['ticker']==tkr].rolling(30).agg({"high": "max", 
                                                "low": "min",
                                                "close": "mean",
                                                "volume": "sum"
                                                })
    roll = roll.dropna()
    print('-'*30,str(tkr),'-'*30)
    print(roll)
```

# 1
- Plot the rolling 10-day minimum of the Facebook closing price using pandas. (Facebook Stock data)

```python
x = fb_stock_prices_2018.rolling(10).agg({"close": "min"}) 
```

```python
plt.plot(x)
plt.title('rolling 10 Days min closing price')
plt.xlabel('data point')
plt.ylabel('min closing price')
plt.show()
```

# 2
- Create a histogram and KDE of the change from open to close in the price of Facebook stock.

```python
fb_stock_prices_2018['change'] = (fb_stock_prices_2018['close']-fb_stock_prices_2018['open'])/fb_stock_prices_2018['open']
```

```python
fb_stock_prices_2018['change'].hist(label='hist',color='cyan')
fb_stock_prices_2018['change'].plot.kde(label='kde',color='red')
plt.legend()
plt.show()
```

# 4
- Make a line plot of the difference between the weekly maximum high price and the weekly minimum low price for Facebook. This should be a single line. (Facebook Stock data).

```python
fb_stock_prices_2018['date'] = pd.to_datetime(fb_stock_prices_2018['date'],errors='coerce')
fb_stock_prices_2018.set_index('date',inplace=True)
plt.figure(figsize=(10,5))
x = fb_stock_prices_2018.resample('W').agg({'low': 'min', 'high': 'max'})
x['change'] = x['high'] - x['low']
x['change'].plot(color='green')
plt.xlabel('date')
plt.ylabel('change')
plt.show()
```

# 5
- Plot the 10-day moving average of the daily change in new COVID-19 cases in Brazil, China, India, Italy, Spain, and the USA. (COVID data)

```python
import warnings
warnings.filterwarnings('ignore')
plt.figure(figsize=(10,5))
for cnt in ['Brazil', 'China', 'India', 'Italy', 'Spain','USA']:
    x = covid19_cases[covid19_cases['countriesAndTerritories']==cnt]
    x['change'] = x['cases'].pct_change()
    x.reset_index(drop=True,inplace=True)
    x['change'].plot(label=str(cnt))
    plt.legend()
plt.show()
```

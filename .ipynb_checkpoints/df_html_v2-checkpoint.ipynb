```python
import pandas as pd
url = 'http://bexar.trueautomation.com/clientdb/Property.aspx?cid=110&prop_id=525041'
df = pd.read_html(url)
```

# Property

```python
df_property = df[0]

columns = df_property.iloc[:,0].to_list() + df_property.iloc[:,2].to_list()
row_value = df_property.iloc[:,1].to_list() + df_property.iloc[:,3].to_list()

df_property = pd.DataFrame([],columns=columns)
df_property.loc[len(df_property)] = row_value
df_property
```

# Values

```python
df_values = df[1]
df_values = df_values[1:].iloc[:,[0,1,2,3]].fillna('')
df_values
```

# Roll History

```python
df_rollhistory = df[11]
df_rollhistory
```

# Taxing Jurisdiction

```python
other_header = df[2]
other_header = other_header.iloc[:,[0,1]]
other_header
```

```python
df_tax = df[3]
df_tax = df_tax.iloc[:,[0,1,2,3,4,5]]
df_tax = df_tax.fillna('')
df_tax
```

# Land

```python
df_land = df[10]
df_land
```

# Save to Excel

```python
with pd.ExcelWriter('output.xlsx') as writer:  
    df_property.to_excel(writer, sheet_name= 'Property',index=False)
    df_values.to_excel(writer, sheet_name= 'Values',index=False)
    other_header.to_excel(writer, sheet_name= 'OtherHeader',index=False)
    df_tax.to_excel(writer, sheet_name= 'Tax',index=False)
    df_land.to_excel(writer, sheet_name= 'Land',index=False)
    df_rollhistory.to_excel(writer, sheet_name= 'RollHistory',index=False)
```

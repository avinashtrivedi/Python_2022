```python
import random
class Boat:
    def __init__(self,boat_name,top_speed=3):
        self.boat_name = boat_name
        self.top_speed = top_speed
        self.current_progress = 0
        
    def set_top_speed(self,new_speed):
        self.top_speed = new_speed
        
    def set_boat_name(self,New_name):
        self.boat_name = New_name
        
    def set_current_progress(self,new_progress):
        self.current_progress = new_progress
    
    def get_boat_name(self):
        return self.boat_name
    
    def get_top_speed(self):
        return self.top_speed
    def get_current_progress(self):
        return self.current_progress
    
    def __str__(self):
        return f'{self.boat_name}: {self.current_progress}'
    
    def move(self):
        random_val = random.randint(0,self.top_speed)
        self.current_progress = self.current_progress + random_val
        return random_val
```

```python
class BoatRace:
    def __init__(self,filename):
        fp = open("the_big_one.csv")
        data = fp.readlines()
        fp.close()
        data = [i.strip().split(',') for i in data]
        
        self.race_name = data[0][1]
        self.race_id = int(data[1][1])
        self.distance = int(data[2][1])
        
        self.racers = []
        for boat_name,t_speed in data[3:]:
            obj = Boat(boat_name,int(t_speed))
            self.racers.append(obj)
        
    def get_race_name(self):
        return self.race_name
    def get_race_id(self):
        return self.race_id
    def get_distance(self):
        return self.distance
    def get_racers(self):
        return self.racers
    
    def add_racer(self,b_obj):
        self.racers.append(b_obj)
        
    def print_racers(self):
        for obj in self.racers:
            print(obj)
            
    def count(self):
        return len(self.racers)
    
    def race(self):
        
        curr_more_than_dist = []
        while len(curr_more_than_dist)==0:
            for racer in self.racers:
                racer.move()
            
            self.print_racers()
        
            for racer in self.racers:
                if racer.current_progress >= self.distance:
                    curr_more_than_dist.append(racer)
                    
        return curr_more_than_dist
```

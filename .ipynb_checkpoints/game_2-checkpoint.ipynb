```python
# 'Popular_Baby_Names.csv'
def babaynames(filename): 
    file = open(filename,'r')
    data = file.readlines()
    file.close()
    names = [line.split(',')[3].strip() for line in data[1:]]
    
    mydict = {}
    for name in names:
        if name[0] not in mydict:
            mydict[name[0]] = [name.title()]
        else:
            mydict[name[0]].append(name.title())
            
    for k in list(mydict):
        mydict[k] = set(mydict[k])
    
    return mydict
```

```python
names = babaynames('Popular_Baby_Names.csv')
```

```python
import random
def lucky3Game():
    print("Welcome to the lucky 3 game")
    total = 0
    while True:
        amount = input('How much do you want to bet or X to exit: ')
        if amount == 'X':
            break
        rolls = []
        for i in range(5):
            rolls.append(random.randint(0,6))
        print("Rolls: ",rolls)

        if rolls.count(3) == 2:
            total = total + float(amount)
            print(f'You Win! Total Winning or loss: ${total}')
        else:
            total = total - float(amount)
            print(f'You lose! Total Winning or loss: ${total}')
```

```python
lucky3Game()
```

```python
import random

def nonsense(nouns,verbs,adjectives,sentence):
    sentence = sentence.split()
    for x in range(len(sentence)):
        if sentence[x] == 'VVV':
            sentence[x] = verbs[random.randrange(0, len(verbs)-1)]
        elif sentence[x] == 'AAA':
            sentence[x] = adjectives[random.randrange(0, len(adjectives)-1)]
        elif sentence[x] == 'NNN':
            sentence[x] = nouns[random.randrange(0, len(nouns)-1)]
    return ' '.join(sentence)
```

```python
nouns= ('action items','applications','bandwidth','deliverables','infrastructures','markets','products','niches','schemas','synergies')
verbs=('architect','e-enable','facilitate','harness','incentivize','leverage','optimize','synergize','unleash','whiteboard')
adj= ('24/7', 'collaborative', 'cutting-edge', 'dynamic', 'end-to-end', 'granular', 'intuitive', 'mission-critical', 'plug-and-play', 'seamless')
text='Our consultants can help you VVV AAA NNN'
```

```python
nonsense(nouns,verbs,adj,text)
```

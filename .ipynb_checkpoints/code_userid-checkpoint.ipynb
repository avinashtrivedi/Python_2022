```python
import pandas as pd
from collections import Counter
```

```python
df = pd.read_csv('data_2.csv')
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp']) 
```

```python
df['month'] = df['timestamp'].dt.month
df['day'] = df['timestamp'].dt.day
```

```python
def get_week(x):
    if 1<=x<=7:
        return 1
    elif 8<=x<=14:
        return 2
    elif 15<=x<=21:
        return 3
    else:
        return 4
```

```python
df['week'] = df['day'].apply(lambda x: get_week(x))
```

```python
# import pickle
ch= "y"
f=open("phonebook.data", "wb")
while ch== "y":
    pn = input("phone number : ")
    nm = input("name : ")
    ei = input("email id : ")
    pc = input("pin code : ")
    db = input( "date of birth : ")
    data = pn + " " + nm + " " + ei + " " + pc + " " + db + "\n"
    ch= input("Do you wanna enter another data (y/n) : ")
    
    f.write(data.encode())
    
    if ch.lower()=="n":
        break
        
f.close()
```

```python
import pickle
pn=input("enter the phone number : ")
try:
    with open("phonebook.data","rb") as f:
        while True:
            try:
                d=pickle.load(f)
                if d.split()[0] == pn:
                    print("phone number ",d.split()[0],"name ",d.split()[1],"email id ",d.split()[3], "pincode ", d.split()[3], "birthday", d.split()[4])
            except:
                print('error')
except EOFError:
    pass
```

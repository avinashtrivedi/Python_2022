# Q2

```python
df_discharge = pd.read_csv("D:\OneDrive - NITT\Custom_Download\discharge_data.csv")
```

```python
df_discharge.head()
```

```python
df_discharge['date'] = pd.to_datetime(df_discharge['date'])
```

```python
df_discharge.set_index('date', inplace=True)
```

```python
df_discharge.head()
```

```python
#A
x = df_discharge.resample('W').mean()
plt.figure(figsize=(10,5))
plt.plot(x,color='green')
plt.xlabel('Weekly Date')
plt.ylabel('Mean')
plt.show()
x
```

```python
#B
x = df_discharge.resample('Q').sum()
x.reset_index(inplace=True)
x['date'] = x['date'].astype('str')
plt.figure(figsize=(10,5))
plt.bar(x['date'],list(x['Num_discharged']),color='g')
plt.xticks(rotation=90)
plt.xlabel('Quarters')
plt.ylabel('Sum')
plt.show()
x
```

# Q3

```python
 df_blood = pd.read_csv("D:\OneDrive - NITT\Custom_Download\Blood_presure.csv")
```

```python
df_blood.head()
```

```python
df_list = []
for ID in set(df_blood['ID']):
    filter_df = df_blood[df_blood['ID'] == ID]
    x = filter_df.to_dict(orient='split')['data']
    x1 = [i for i in x if 'Diastolic Blood Pressure' in i]
    x2 = [i for i in x if 'Systolic Blood Pressure' in i]
    
    for k1,k2 in zip(x1,x2):
        d = {}
        d['ID'] = k1[0]
        d['Diastolic Blood Pressure'] = k1[2]
        d['Systolic Blood Pressure'] = k2[2]
        df_list.append(d)

    if len(x1)>len(x2):
        n = len(x1) - len(x2)
        x1 = x1[-n:]
        for k1 in x1:
            d = {}
            d['ID'] = k1[0]
            d['Diastolic Blood Pressure'] = k1[2]
            d['Systolic Blood Pressure'] = float('NaN')
            df_list.append(d)
    elif len(x1)<len(x2):
        n = len(x2) - len(x1)

        x2 = x2[-n:]

        for k2 in x2:
            d = {}
            d['ID'] = k2[0]
            d['Diastolic Blood Pressure'] = float('NaN')
            d['Systolic Blood Pressure'] = k2[2]
            df_list.append(d)
```

```python
df_blood_new = pd.DataFrame(df_list)
```

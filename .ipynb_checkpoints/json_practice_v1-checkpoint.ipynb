```python
resultdata = data[5]
```

```python
full = pd.DataFrame([])
for resultdata in data:
    for i in range(len(resultdata['annotations'][0]['result'])):
        x = resultdata['annotations'][0]['result'][i]
        temp = {k:x[k] for k in x if k in ('id','origin','type')}
        temp['start'] = x['value']['start']
        temp['end'] = x['value']['end']
        df = pd.json_normalize(temp)
        full = full.append(df)
full.reset_index(drop=True)
```

```python
import pandas as pd
# import pandas as pd
import os
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
def backtesting(pattern):
    data = []
    for fname in os.listdir('BOSCHLTD'):
        path = os.path.join('BOSCHLTD',fname)
        df = pd.read_csv(path)

        df = df[['Date','Symbol',pattern,'Close']]
        for n in [5,10,15]:
            x = df['Close'].pct_change(n)*100
            df[f'return_{n}'] = list(x[n:]) + [float('NaN')]*n
        df = df[:-5]
        df = df[df[pattern]!=0]
        data.append(df)
    pd.concat(data).to_csv(pattern+'.csv')
```

```python
backtesting('CDLDOJI')
```

```python
def generate_activity(x):
    print(type(x),'-->',x['Delivery'])
    if ((x['Delivery'] > x['Deliverble_mean']) & (x['Action'] > x['Action_mean'])):
        return 'Jackpot'
    elif ((x['Delivery'] > x['Deliverble_mean']) & (x['Action'] < x['Action_mean'])):
        return 'Accumalation'
    elif ((x['Delivery'] < x['Deliverble_mean']) & (x['Action'] > x['Action_mean'])):
        return 'Action'
    elif ((x['Delivery'] < x['Deliverble_mean']) & (x['Action'] < x['Action_mean'])):
        return 'No Interest'
    return ''

df['Activity'] = df.apply(generate_activity, axis=1)
```

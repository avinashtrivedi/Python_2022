```python
import numpy as np

def sphere_fun(r):
    # x^2 + y^2
    return r[0]*r[0]+r[1]*r[1]

def grad_sphere_fun(r):
    # [2*x,2*y]
    return np.array([2*r[0],2*r[1]])

def Rosenbrock_fun(r):
    # 100*(y-x**2)**2 + (1-x)**2
    return 100*(r[1]-r[0]**2)**2 + (1-r[0])**2

def grad_Rosenbrock_fun(r):
    # [2*100*(y-x**2)*(-2*x) + 2*(1-x)*(-1), 2*100*(y-x**2)]
    return np.array([2*100*(r[1]-r[0]**2)*(-2*r[0]) + 2*(1-r[0])*(-1), 2*100*(r[1]-r[0]**2)])

def Himmelblaus_fun(r):
    # (x**2+y-11)**2 + (x + y**2-7)**2
    return (r[0]**2+r[1]-11)**2 + (r[0] + r[1]**2 - 7)**2

def grad_Himmelblaus_fun(r):
    # [2*(x**2+y-11)*2*x + 2*(x + y**2-7), 2*(x**2+y-11) + 2*(x + y**2-7)*2*y]
    return np.array([2*(r[0]**2+r[1]-11)*2*r[0] + 2*(r[0] + r[1]**2-7), 2*(r[0]**2+r[1]-11) + 2*(r[0] + r[1]**2-7)*2*r[1] ])

def grad_descent(ri, f, grad_f, s, tolerance, max_iterations):
    # ri = initial guess
    # one step (s=step size):
    # r = r - s*grad(f)
    ri = np.array(ri)
    r = ri
    num_step = 0
    for i in range(max_iterations):
        #e = s*grad(r,f,s)
        e = s*grad_f(r)
        em = np.sqrt(np.dot(e,e))
        if( em > 1 ):
            e = e / em
        if( em < tolerance ):
            break;
        r = r - e
        #print(f"r({i}) = ",r, " grad = ",e/s)
        num_step += 1
    return r, num_step



ri = [1,1]
max_iterations = 50
print("sphere fun.:\nri = ",ri)
rf, num_it = grad_descent(ri, sphere_fun, grad_sphere_fun, 0.2, 0.0001,  max_iterations)
print("rf = ",rf, "number of iterations = ",num_it, " out of (",max_iterations,")")

ri = [0.5,2]
max_iterations = 100
print("Rosenbrock fun\nri = ",ri)
rf, num_it = grad_descent(ri, Rosenbrock_fun, grad_Rosenbrock_fun, 0.0001, 0.0000001,  max_iterations)
print("rf = ",rf, "number of iterations = ",num_it, " out of (",max_iterations,")")


ri = [2,2]
max_iterations = 100
print("Himmelblaus fun\nri = ",ri)
rf, num_it = grad_descent(ri, Himmelblaus_fun, grad_Himmelblaus_fun, 0.01, 0.0001,  max_iterations)
print("rf = ",rf, "number of iterations = ",num_it, " out of (",max_iterations,")")
```

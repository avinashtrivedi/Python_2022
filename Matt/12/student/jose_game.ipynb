```python
import random

# Main function
# 'main' logic of the program
def main():

    # while the user keeps entering 'y'
    while True:
        # display game header
        game_header()
        # prompt user to enter name
        name = input(">Enter Your Name:")
        # prompt user for rounds
        round_play = int(input(">Enter number of rounds to play:"))
        
        print('\n')
        # initialize computer score to 0
        computer_score = 0
        # initialize user score to 0
        user_score = 0

        # loop for number of rounds user wants to play
        for num in range(round_play):
            # print round #
            print(f"Round #{num+1}")
            print('-'*8)
            # call rock paper scissors function (it returns values to you)
            c_score,u_score = rock_paper_scissors()
            # update computer score
            computer_score = computer_score + c_score
            # update user score
            user_score = user_score + u_score
            
            # display scoreboard
            score_board(computer_score, user_score,name)
            
        if computer_score>user_score:
            print("TEST: 1 0")
            print('Game Over! I win :)')
        elif computer_score<user_score:
            print("TEST: 0 1")
            print('Game Over! you win :)')
        else:
            print("TEST: 0 0")
            print('Game Over! Its Draw :)')
        # call continue_game function (it returns values to you)
        decision = continue_game()
        if decision==False:
            break
            
# Generates a random choice for the computer
# choices: rock, paper, scissors
def generate_computer_choice():
    rn = random.randint(1,3) # generate a random number
    
    # use the random number to return 'rock', 'paper', or 'scissors
    if rn==1:
        return 'rock'
    elif rn==2:
        return 'paper'
    else:
        return 'scissors'

# RPS functionality
# returns the scores for computer and user 
def rock_paper_scissors():
    # display game header
    print("[Choose Rock,Paper,Scissors]")
    # prompt the user to enter a choice
    c_user = input(">Enter choice: ")
    print('\n')
    # call 'generate_computer_choice' function to generate choice for computer
    c_comp = generate_computer_choice()
    
    print('-'*45)
    # if statements for the game start here
    if c_user.lower() == 'rock' and c_comp == 'scissors':
        print(f"You picked {c_user} i picked {c_comp} and I Lost!")
        print('-'*45)
        return (0,1) # rock  (computer,user)
    elif c_user.lower() == 'scissors' and c_comp == 'paper':
        print(f"You picked {c_user} i picked {c_comp} and I Lost!")
        print('-'*45)
        return (0,1) # scissors
    elif c_user.lower() == 'paper' and c_comp == 'rock':
        print(f"You picked {c_user} i picked {c_comp} and I Lost!")
        print('-'*45)
        return (0,1) # paper
    
    elif c_comp == 'rock' and c_user.lower() == 'scissors':
        print(f"You picked {c_user} i picked {c_comp} and I Win!")
        print('-'*45)
        return (1,0) # rock
    elif c_comp == 'scissors' and c_user.lower() == 'paper':
        print(f"You picked {c_user} i picked {c_comp} and I Win!")
        print('-'*45)
        return (1,0) # scissors
    elif c_comp == 'paper' and c_user.lower() == 'rock':
        print(f"You picked {c_user} i picked {c_comp} and I Win!")
        print('-'*45)
        return (1,0)
    else:
        print(f"You picked {c_user} i picked {c_comp} and Its a Draw!")
        print('-'*45)
        return (0,0)
        
# Displays the header of the game
def game_header():
    print("** Rock Paper SCISSORS **")# display game header
    print("[Press any key to start  the game]\n")# prompt user to press enter to start

# Displays the scores for computer and user
# You will need to pass three arguments to this function
def score_board(computer_score, user_score,user_name):
    print("\n<SCOREBOARD>")
    print('-'*13)
    # Display computer score
    print('Computer:',computer_score)
    
    # Display user score
    print(f'{user_name}:',user_score)
    print('\n')

# Prompts the user to continue or not
# Options: Y/Q
def continue_game():
    # ask user if they want to continue
    c = input("Play again? Enter 'Y' for Yes,'Q' to quit the game: ")
    # if they do, return true
    if c=='Y':
        return True
    elif c=='Q':
        return False
    # if they don't, return false

main()
```

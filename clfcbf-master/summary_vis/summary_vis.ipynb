```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter
font = {'size': 32}
matplotlib.rc('font', **font)

colors = {'ours': '#8c1515', 'realsyn': '#b26f16', 'mamps': '#d2c295', 'pic': '#dad7cb', 'maddpg': '#b6b1a9'}
alphas = {'ours': 0.7, 'realsyn': 0.6, 'mamps': 1.0, 'pic': 1.0, 'maddpg': 1.0}
error_kw = {'elinewidth': 4, 'capthick': 4, 'capsize': 10, 'ecolor':  'black', 'alpha': 0.6}
```

```python
fig = plt.figure(figsize=(16, 10))
ax = fig.add_subplot(111)

width = 0.17
ax.bar(np.arange(5) - width * 0.5, 
       [0.9999, 0.9999, 0.9985, 0.9999, 0.9985], 
        width,
        yerr=[0, 0, 0, 0, 0],
        error_kw=error_kw,
        color=colors['ours'], 
        alpha=alphas['ours'],
        edgecolor='grey',
        linewidth=3,
        label='MDBC (Ours)')

ax.bar(np.arange(5) + width * 0.5, 
       [0.9992, 0.9550, 0.9107, 0.9550, 0.9107], 
        width,
        yerr=[0.0002, 0.009, 0.012, 0, 0],
        error_kw=error_kw,
        color=colors['mamps'], 
        alpha=alphas['mamps'],
        edgecolor='grey',
        linewidth=3,
        label='MAMPS')


ax.set_xticks([0, 1, 2, 3, 4])
ax.set_xticklabels(['2D Quadrotor', '3D Quadrotor', 'Neural Lander', 'Satellite Rendezvous', 'Segway'])

ax.set_xlabel('Number of agents')
ax.set_ylabel('Safety rate', fontsize=36)

ax.set_ylim((0.89, 1.01))
for side in ax.spines.keys():
    ax.spines[side].set_linewidth(2)
    ax.spines[side].set_color('grey')
    
plt.title('Maze')

plt.legend(loc='lower left', framealpha=0.8, facecolor='white')
plt.tight_layout()
plt.savefig('./maze_safety.png')
plt.show()
```

# Q2.

```python
df = pd.read_csv('healthcare-spending.csv',skiprows=2)
df = df.loc[1:51]
```

```python
X = [int(col.split('__')[0]) for col in list(df)[1:]]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6,15))
for i in range(len(df)):
    state = list(df.iloc[i])[0]
    spending = list(df.iloc[i])[1:]
    plt.plot(X,spending,label=state)
    
plt.xlabel('Year')
plt.ylabel('Spending')
plt.legend()
plt.show()
```

```python
state = []
spending = []
for i in range(len(df)):
    state.append(list(df.iloc[i])[0])
    spending.append(sum(list(df.iloc[i])[1:])/len(df.iloc[i][1:]))
```

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_axes([0,0,1,1])
ax.bar(state,spending,color='green')
plt.xticks(rotation=90)
plt.show()
```

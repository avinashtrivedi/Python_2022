```python
def check_errors(pn,time,ft,lp,fa):
    
    errors = []
    if 1<=int(pn)<=8:
        pass
    else:
        #An error in the pump number field
        errors.append('Invalid pump number.')

    if 0<=int(time[:2])<=23 and 0<=int(time[2:]) <=59:
        pass
    else:
        #An error in the time field
        errors.append('Invalid time.')

    if ft in ('E','U','P','D'):
        pass
    else:
        #An error in the fuel type field
        errors.append('Invalid fuel type.')

    if float(lp)>=0:
        pass
    else:
        #An error in the Litre price field
        errors.append('Invalid Litre price.')

    if float(fa)>=2:
        pass
    else:
        #An error in the Fill amount field
        errors.append('Invalid Fill amount.')
    return errors

# list to store data as list of list
data = []

# assign serial number to data
sr = 1

print("""UTOPIA PETROL STATIONS SYSTEM
-----------------------------
""")

# iterate through the loop to keep getting the Main Menu 
while True:
    
    # try-except block to handle exception.
    try:
        
        print("""
Main menu
---------
[A]dd a new entry
[L]ist entries
[D]elete an entry
[S]ave entries
[Q]uit""")
        choice = input("Enter choice: ")
        
        # receive inputs to Add a new entry
        if choice == 'A':
            print("\nAdd a new entry")
            print("---------------")
            pn = input("Pump number [1-8]: ") 
            time = input("Time HHMM: ")
            ft = input("Fuel Type [E]94, [U]91, [P]98, [D]iesel]: ")
            lp = input("Litre price: ")
            fa = input("Fill amount: ")

            # check the conditions
            check = check_errors(pn,time,ft,lp,fa)

            # store the data if its valid
            if not check:
                data.append([sr,pn,time,ft,lp,fa])
                sr = sr + 1
                print("** Entry added **\n")
            else:
                for i in check:
                    print('-',i)

        # list the data
        elif choice == 'L':
            print('\n')
            print("{:<7} {:<7} {:<5} {:<7} {:<7} {:<7}".format('EntryNo','PumpNo','Time','FType','LPrice','FAmount'))
            print('-'*45)
            for i in data:
                print("{:<7} {:<7} {:<5} {:<7} {:<7} {:<7}".format(i[0],i[1],i[2],i[3],i[4],i[5]))

            print('\n')

        # delete the data
        elif choice == 'D':
            print("\n{:<7} {:<7} {:<5} {:<7} {:<7} {:<7}".format('EntryNo','PumpNo','Time','FType','LPrice','FAmount'))
            for i in data:
                print("{:<7} {:<7} {:<5} {:<7} {:<7} {:<7}".format(i[0],i[1],i[2],i[3],i[4],i[5]))
            print('\n')
            entry = int(input("Enter the number of the entry to be deleted: "))

            # iterate through the data and delete the specific entry
            for i in range(len(data)):
                if data[i][0] == entry:
                    _ = data.pop(i)
                    print(f"** Entry {entry} deleted **\n")
                    break
        
        # Quit/
        elif choice == 'Q':
            break
            
        # Save the data in a file
        elif choice == 'S':
            print('\nSave entries')
            print('------------\n')
            name = input("Enter the file name: ")

            # create a file pointer
            fp = open(name,'w')
            newline = ''
            
            # iterate through the data and write into the file
            for i in data:
                
                # write the newline
                fp.write(newline)
                
                # write the data
                fp.write("{},{},{},{},{}".format(i[1],i[2],i[3],i[4],i[5]))
                newline = '\n'
                
            # close the file
            fp.close()

            print(f"\n** Data saved to {name} **")
        else:
            print('Wrong choice. Please try again.')
    except:
        print('Wrong Inputs.')
```

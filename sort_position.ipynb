```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')


# In[2]:


n = 3
df = pd.read_csv('sample.csv')
df.dropna(inplace=True)
df['DATE'] = pd.to_datetime(df['DATE'])
th = sorted(set(df['DATE']))[-n]
df_new = df[df['DATE']>=th]

d = df_new.groupby(['SYMBOL']).agg(
    {
'activity' : lambda x: set(x),  
    }
)

d = d.reset_index(level=0)
d['flag'] = d['activity'].apply(lambda x: 1 if len(x)==1 else 0)
d = d[d['flag']==1]
d['activity'] = d['activity'].apply(lambda x: x.pop())
d.drop('flag',axis=1,inplace=True)
d.reset_index(drop=True,inplace=True)


# In[3]:


d


# In[ ]:




```

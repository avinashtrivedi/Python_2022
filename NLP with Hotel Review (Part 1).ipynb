#### Creating numeric columns for certain `Tags` 

```python
# get count of each tag
tagcount = hotelreviews_data2['Tags']
pd.set_option('display.max_rows', None)
tagcount.str.split(',').explode().value_counts()
```

We have the top tags in the count above.

Given this, we can create numeric columns for the following tags:
<ul>
<li>Leisure trip (0/1)</li>
<li>Business trip (0/1)</li>
<li>Submitted from a mobile device (0/1)</li>
<li>Number of nights stayed</li>
<li>Used double room (0/1)</li>
<li>Couple (0/1)</li>
<li>Solo traveler (0/1)</li>
<li>Group (0/1)</li>
<li>Family with children (0/1)</li>
<li>Friends (0/1)</li>
</ul>

```python
# create 'Leisure trip' column and set binary mapping
hotelreviews_data2['Leisure trip'] = pd.np.where(hotelreviews_data2['Tags'].str.contains("Leisure trip"), 1, 0)

# create 'Business trip' column and set binary mapping
hotelreviews_data2['Business trip'] = pd.np.where(hotelreviews_data2['Tags'].str.contains("Business trip"), 1, 0)

# create 'Submitted from mobile device' column and set binary mapping
hotelreviews_data2['Submitted from mobile device'] = pd.np.where(hotelreviews_data2['Tags'].str.contains("Submitted from a mobile device"), 1, 0)

# create 'Used double room' column and set binary mapping. 
# Note that this is set up to pick up any tag that contains 'Double' so as to capture all Double room tags.
hotelreviews_data2['Used double room'] = pd.np.where(hotelreviews_data2['Tags'].str.contains("Double"), 1, 0)

# create 'Group: Couple' column and set binary mapping
hotelreviews_data2['Group: Couple'] = pd.np.where(hotelreviews_data2['Tags'].str.contains("Couple"), 1, 0)

# create 'Group: Solo Traveler' column and set binary mapping
hotelreviews_data2['Group: Solo Traveler'] = pd.np.where(hotelreviews_data2['Tags'].str.contains("Solo traveler"), 1, 0)

# create 'Group: Family with children' column and set binary mapping. 
# Note that this is set up to pick up any tag that contains 'Family with' so as to capture all instances of 'Family with younger children' and 'Family with older children' tags.
hotelreviews_data2['Group: Family with children'] = pd.np.where(hotelreviews_data2['Tags'].str.contains("Family with"), 1, 0)

# create 'Group: Friends' column and set binary mapping
hotelreviews_data2['Group: Friends'] = pd.np.where(hotelreviews_data2['Tags'].str.contains("Friends"), 1, 0)
```

```python
# create 'Number of nights stayed' column and set mapping
texts = [f"Stayed {i} nights" for i in range(1, 32)]
num_of_nights = []
for i in range(len(hotelreviews_data2)):
    for j in range(31):
        if texts[j] in hotelreviews_data2['Tags'][i]:
            num_of_nights.append(j+1)
            break
    else:
        num_of_nights.append(0)
hotelreviews_data2['Number of nights stayed'] = num_of_nights
```

### D: Use a count vectorizer to combine `Positive_Review` and `Negative_Review` with the numeric data (notice that this is done AFTER the train/test split). You should vectorize each column separately, ending up with two sparse matrixes, and then combine the three matrixes (numeric data, positive matrix, negative matrix). You may have to adjust the min_df parameter.

```python
# import countvectorizer
from sklearn.feature_extraction.text import CountVectorizer
```

We will now vectorize `Positive_Review`

```python
# instantiate
bagofwords = CountVectorizer()

# fit
bagofwords.fit(hotelreviews_data3['Positive_Review'])

# transform
positive_review_transformed = bagofwords.transform(hotelreviews_data3['Positive_Review'])
positive_review_transformed
```

We will then vectorize the `Negative_Review` column.

```python
# instantiate Negative_Review column
bagofwords = CountVectorizer()

# fit
bagofwords.fit(hotelreviews_data3['Negative_Review'])

# transform
negative_review_transformed = bagofwords.transform(hotelreviews_data3['Negative_Review'])
negative_review_transformed
```

Now, we will create a matrix out of the dataframe

```python
from scipy.sparse import coo_matrix, hstack
```

```python
A = coo_matrix(positive_review_transformed)
B = coo_matrix(negative_review_transformed)
```

```python
Z = hotelreviews_data3[:5].dtypes
```

```python
d_type_indx = [i for i,typ in enumerate(list(Z)) if typ in ('int32','int64','float64')]
```

```python
numeric_data = hotelreviews_data3[Z.index[d_type_indx]]
```

```python
combined_data = hstack([A,B,numeric_data])
```

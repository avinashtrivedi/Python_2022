```python
# Import packages
import numpy as np #deals with arrays
import matplotlib.pyplot as plt #plotting package
from pylab import *
import netCDF4
import pandas as pd
### Opening with xarray to do time selections
import xarray as xr #deals with multidimensional arrays
data = xr.open_dataset('/Volumes/WorkDrive/melt_dates/seaiceconc.nc') # opening sit for xarray
## Arange months up to November and December ##
def is_months(month):
    return (month >= 7) & (month <= 12)
data = data.sel(time=is_months(data['time.month']))

## Select SIC variable ##
sic_xr = data.SI_12km_SH_ICECON_DAY_SpPolarGrid12km
latitude = data.GridLat_SpPolarGrid12km
longitude = data.GridLon_SpPolarGrid12km

## Here we want all the values that are over 100 (missing or land mask) to be set to nan ##
sic_xr = sic_xr.where(sic_xr < 101,np.nan)

## Because the desired output is one slice/ image/grid for each year, I will process year by year


da_2012_only = sic_xr.sel(time=sic_xr.time.dt.year.isin([2012]))
da_2013_only = sic_xr.sel(time=sic_xr.time.dt.year.isin([2013]))
da_2014_only = sic_xr.sel(time=sic_xr.time.dt.year.isin([2014]))
da_2015_only = sic_xr.sel(time=sic_xr.time.dt.year.isin([2015]))
da_2016_only = sic_xr.sel(time=sic_xr.time.dt.year.isin([2016]))
da_2017_only = sic_xr.sel(time=sic_xr.time.dt.year.isin([2017]))
da_2018_only = sic_xr.sel(time=sic_xr.time.dt.year.isin([2018]))
da_2019_only = sic_xr.sel(time=sic_xr.time.dt.year.isin([2019]))
da_2020_only = sic_xr.sel(time=sic_xr.time.dt.year.isin([2020]))
da_2021_only = sic_xr.sel(time=sic_xr.time.dt.year.isin([2021]))

def continuous_meet(cond, window_size, dim):
    """
    Continuously meet a given condition along a dimension.
    """
    _found = cond.rolling(dim={'time': window_size},
                          center=True).sum(skipna=True).fillna(False).astype(np.float)

    detected = np.array(
        _found.rolling(dim={'time': window_size})
        .reduce(lambda a, axis: (a == window_size).any(axis=axis))
        .fillna(False)
        .astype(bool)
    )

    indices = (detected * np.arange(detected.shape[0]).reshape(detected.shape[0], 1, 1))
    indices[indices == 0] = detected.shape[0]
    output = indices.argmin(axis=0)

    return xr.DataArray(output)

########################################################################################
breakup2012 = continuous_meet(da_2012_only <= 15, window_size=3, dim='time')
breakup2013 = continuous_meet(da_2013_only <= 15, window_size=3, dim='time')
breakup2014 = continuous_meet(da_2014_only <= 15, window_size=3, dim='time')
breakup2015 = continuous_meet(da_2015_only <= 15, window_size=3, dim='time')
breakup2016 = continuous_meet(da_2016_only <= 15, window_size=3, dim='time')
breakup2017 = continuous_meet(da_2017_only <= 15, window_size=3, dim='time')
breakup2018 = continuous_meet(da_2018_only <= 15, window_size=3, dim='time')
breakup2019 = continuous_meet(da_2019_only <= 15, window_size=3, dim='time')
breakup2020 = continuous_meet(da_2020_only <= 15, window_size=3, dim='time')
breakup2021 = continuous_meet(da_2021_only <= 15, window_size=3, dim='time')
########################################################################################

# Turn into data array ###
breakup2012 = xr.DataArray(breakup2012)
breakup2012 = xr.DataArray(
    data=breakup2012,
    dims=["ygrid","xgrid"],
    coords=dict(
        longitude=(["ygrid","xgrid"],longitude),
        latitude=(["ygrid","xgrid"],latitude),
        ),
        attrs=dict(
        description="Sea Ice Breakup",
        ),
)

## 2013 ##
breakup2013 = xr.DataArray(breakup2013)
breakup2013 = xr.DataArray(data=breakup2013,
                     dims=["ygrid","xgrid"],
                     coords=dict(
                         longitudes=(["ygrid","xgrid"],longitude),
                         latitudes=(["ygrid","xgrid"],latitude),
                                ),
                     attrs=dict(description="Sea Ice Breakup",
                               ),
                    )
## 2014 ##
breakup2014= xr.DataArray(breakup2014)
breakup2014 = xr.DataArray(data=breakup2014,
                     dims=["ygrid","xgrid"],
                     coords=dict(
                         longitudes=(["ygrid","xgrid"],longitude),
                         latitudes=(["ygrid","xgrid"],latitude),
                                ),
                     attrs=dict(description="Sea Ice Breakup",
                               ),
                    )
## 2015 ##
breakup2015 = xr.DataArray(breakup2015)
breakup2015 = xr.DataArray(data=breakup2015,
                     dims=["ygrid","xgrid"],
                     coords=dict(
                         longitudes=(["ygrid","xgrid"],longitude),
                         latitudes=(["ygrid","xgrid"],latitude),
                                ),
                     attrs=dict(description="Sea Ice Breakup",
                               ),
                    )
## 2016 ##
breakup2016 = xr.DataArray(breakup2016)
breakup2016 = xr.DataArray(data=breakup2016,
                     dims=["ygrid","xgrid"],
                     coords=dict(
                         longitudes=(["ygrid","xgrid"],longitude),
                         latitudes=(["ygrid","xgrid"],latitude),
                                ),
                     attrs=dict(description="Sea Ice Breakup",
                               ),
                    )
## 2017 ##
breakup2017 = xr.DataArray(breakup2017)
breakup2017 = xr.DataArray(data=breakup2017,
                     dims=["ygrid","xgrid"],
                     coords=dict(
                         longitudes=(["ygrid","xgrid"],longitude),
                         latitudes=(["ygrid","xgrid"],latitude),
                                ),
                     attrs=dict(description="Sea Ice Breakup",
                               ),
                    )
## breakup2018 ##
breakup2018 = xr.DataArray(breakup2018)
breakup2018 = xr.DataArray(data=breakup2018,
                     dims=["ygrid","xgrid"],
                     coords=dict(
                         longitudes=(["ygrid","xgrid"],longitude),
                         latitudes=(["ygrid","xgrid"],latitude),
                                ),
                     attrs=dict(description="Sea Ice Breakup",
                               ),
                    )
## 2019 ##
breakup2019 = xr.DataArray(breakup2019)
breakup2019 = xr.DataArray(data=breakup2019,
                     dims=["ygrid","xgrid"],
                     coords=dict(
                         longitudes=(["ygrid","xgrid"],longitude),
                         latitudes=(["ygrid","xgrid"],latitude),
                                ),
                     attrs=dict(description="Sea Ice Breakup",
                               ),
                    )
## 2020 ##
breakup2020 = xr.DataArray(breakup2020)
breakup2020 = xr.DataArray(data=breakup2020,
                     dims=["ygrid","xgrid"],
                     coords=dict(
                         longitudes=(["ygrid","xgrid"],longitude),
                         latitudes=(["ygrid","xgrid"],latitude),
                                ),
                     attrs=dict(description="Sea Ice Breakup",
                               ),
                    )
## 2021 ##
breakup2021 = xr.DataArray(breakup2021)
breakup2021 = xr.DataArray(data=breakup2021,
                     dims=["ygrid","xgrid"],
                     coords=dict(
                         longitudes=(["ygrid","xgrid"],longitude),
                         latitudes=(["ygrid","xgrid"],latitude),
                                ),
                     attrs=dict(description="Sea Ice Breakup",
                               ),
                    )
## To netcdf file ##
breakup2012.to_netcdf('/Volumes/WorkDrive/melt_dates/files/y12_break3d_30.nc')
breakup2013.to_netcdf('/Volumes/WorkDrive/melt_dates/files/y13_break3d_30.nc')
breakup2014.to_netcdf('/Volumes/WorkDrive/melt_dates/files/y14_break3d_30.nc')
breakup2015.to_netcdf('/Volumes/WorkDrive/melt_dates/files/y15_break3d_30.nc')
breakup2016.to_netcdf('/Volumes/WorkDrive/melt_dates/files/y16_break3d_30.nc')
breakup2017.to_netcdf('/Volumes/WorkDrive/melt_dates/files/y17_break3d_30.nc')
breakup2018.to_netcdf('/Volumes/WorkDrive/melt_dates/files/y18_break3d_30.nc')
breakup2019.to_netcdf('/Volumes/WorkDrive/melt_dates/files/y19_break3d_30.nc')
breakup2020.to_netcdf('/Volumes/WorkDrive/melt_dates/files/y20_break3d_30.nc')
breakup2021.to_netcdf('/Volumes/WorkDrive/melt_dates/files/y21_break3d_30.nc')
```

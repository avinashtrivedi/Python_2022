```python
import numpy as np
import pandas as pd
```

```python
import xml.etree.ElementTree as ET
```

```python
tree = ET.parse('E:/Project/Test/Q22020_35460_20220316.xml')
root = tree.getroot()
```

```python
#Checking all tags
for child in root:
    print(child.tag)
```

```python
#Remove header
root.remove(root[0])
```

```python
for child in root:
    print(child.tag)
```

```python
#Checking tags present in each PatientLevelData tag
for child in root[0]:
    print(child.tag)
```

```python
#Capturing columns for the CSV file
cols = []
for child in root:
    for grandchild in child:
        if grandchild.tag not in cols:
            cols.append(grandchild.tag)
            if(grandchild.tag in ["ANALYSIS","DEMOGRAPHICS","COMMENTS"]):
                text = grandchild.tag.lower() + "_"
                for elem in grandchild[0]:
                    new_text = text + elem.tag.lower()
                    cols.append(new_text)
            if(grandchild.tag == "CAHPS"):
                text = grandchild.tag.lower() + "_"
                cols.append(text + "varname")
                cols.append(text + "value")
                
cols.remove('ANALYSIS')
cols.remove('DEMOGRAPHICS')
cols.remove('COMMENTS')
cols.remove('CAHPS')
```

```python
cols
```

```python
#Reordering the columns
cols1 = cols[:10]
cols2 = cols[10:]
cols2 = cols2[-3:] + cols2[:-3]
cols = cols1+cols2
```

```python
cols
```

```python
#Capturing row wise data for each row in the dataframe
all_cols = []
indices = []
for child in root:
    o_cols = []
    an_cols = []
    de_cols = []
    co_cols = []
    ca_cols = []
    for grandchild in child:
        an_col = []
        de_col = []
        co_col = []
        if(grandchild.tag == "ANALYSIS"):
            for ggc in grandchild:
                for elem in ggc:
                    an_col.append(elem.text)
                an_cols.append(an_col)
                an_col = []
        if(grandchild.tag == "DEMOGRAPHICS"):
            for ggc in grandchild:
                for elem in ggc:
                    de_col.append(elem.text)
                de_cols.append(de_col)
                de_col = []
        if(grandchild.tag == "COMMENTS"):
            for ggc in grandchild:
                for elem in ggc:
                    co_col.append(elem.text)
                co_cols.append(an_col)
                co_col = []
        if(grandchild.tag == "CAHPS"):
            j = 0
        else:
            o_cols.append(grandchild.text)
            
    m = max(len(an_cols),len(de_cols),len(co_cols),len(ca_cols))
    
    i = 0
    while i<m:
        indices.append(i)
        new_col = []
        new_col = new_col + o_cols[:5]
        if i>=len(an_cols):
            new_col = new_col + ["",""]
        else:
            new_col = new_col + an_cols[i]
        if i>=len(de_cols):
            new_col = new_col + ["",""]
        else:
            new_col = new_col + de_cols[i]
        if i>=len(co_cols):
            new_col = new_col + ["","",""]
        else:
            new_col = new_col + co_cols[i]
            
        new_col = new_col + ["",""]
        
        all_cols.append(new_col)
        
        i = i+1
```

```python
all_cols
```

```python
#Create the dataframe
df = pd.DataFrame(all_cols,index = indices,columns = cols)
```

## Modeling

### Q1: Employ a linear classifier on this dataset:

### A: Fit a logistic regression model to this data with the solver set to `lbfgs`. What is the accuracy score on the test set?

We will first set up X and y variables based on `rating` being the target column on the training set first.

```python
X_train = train_df.drop(columns="rating")
y_train = train_df["rating"]
```

```python
X_train.shape
```

```python
y_train.shape
```

Now that we have created the training set, we will fit a logistic regression  with solver set to `lbfgs`.

```python
# instantiate model
train_logit = LogisticRegression(solver='lbfgs', random_state=1)

#fit model
train_logit.fit(X_train, y_train);
```

```python
train_logit.intercept_
```

```python
train_logit.coef_
```

```python
# score model
print(f'Score on train: {train_logit.score(X_train, y_train)}')
```

We will now fit the logistic regression with solver set to `lbfgs` to the test set. Like the training set, the test set's variables will be based on `rating` being the target column.

```python
X_test = test_df.drop(columns="rating")
y_test = test_df["rating"]
```

```python
# instantiate model
test_logit = LogisticRegression(solver='lbfgs', random_state=1)

#fit model
test_logit.fit(X_test, y_test);
```

```python
# score model
print(f'Score on test: {test_logit.score(X_test, y_test)}')
```

The accuracy score on the test set is 0.7190063276306539. This is lower than the accuracy score on the training set which is at 0.7238631036099391.

```python
import pandas as pd
import os
from tqdm.notebook import tqdm
```

```python
path = "D:\OneDrive - NITT\Custom_Download\dataset_onedrive" # modify here
```

```python
csfiles = os.path.join(path,'04_final_output_after_name_logic_adjustment')
```

```python
csfiles
```

```python
os.listdir(csfiles)
```

```python
all_csv_paths = []
for file in os.listdir(csfiles):
    full_path = os.path.join(csfiles,file)
    for csv in os.listdir(full_path):
        temp_path = os.path.join(full_path,csv)
        if csv.split('.')[1] in ('csv','xlsx'):
            all_csv_paths.append(temp_path)
```

```python
len(all_csv_paths)
```

```python
all_csv_paths[0]
```

```python
all_df = []
for csv in tqdm(all_csv_paths):
    try:
        if csv[-3:]=='lsx':
            df = pd.read_excel(csv)
        else:
            df = pd.read_csv(csv)

        all_df.append(df)
    except:
        print('failed:',csv)
```

```python
# df = pd.concat(all_df)
```

```python
df = all_df[0].copy()
count = 0
for temp_df in tqdm(all_df):
    try:
        temp = temp_df[['legislature','session','date','speech_order', 'speaker','chair','position','speech','terms']]
        df = df.append(temp)
        count = count + 1
    except:
        print('failed',count)
```

```python
df
```

```python
df.to_csv('df_full.csv',index=False)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from glob import glob
import re
import os
from pylab import rcParams
rcParams['figure.figsize'] = 10,7
folder = 'Saliva_20_ul'
folder += os.sep
cor_wavelength = 650
mes_wavelength = 450

def load_tsv(filename):
    df = pd.read_csv(filename, sep='\t', header=[0,1])

    idx = np.where(df.index.get_loc('//QSpecEnd: ') == True)[0] + 1
    idx = np.append(idx, len(df)+10)
    spectrums = {}
    for i in range(len(idx)-1):
        start = idx[i]
        end =  idx[i+1]-10
        if i > 0:
            name = df.iloc[start-10].name
        else:
            name = df.columns.get_level_values(0)[0]
        temp = df.iloc[start:end].astype('float')
        temp.index = temp.index.astype('float')
        spectrums[name] = temp

    df = pd.concat(spectrums, axis=1)
    df.columns = df.columns.droplevel([1,2])
    return df

test = glob(folder + '*.tsv')[0]
load_tsv(test).plot()

experiments = {}
for filename in glob(folder + '*.tsv'):
    if 'table.tsv' not in filename:
        experiments[filename.strip(folder[:-1])[1:-4]] = load_tsv(filename)

experiments = pd.concat(experiments, axis=1)

corrected = experiments.subtract(experiments.loc[cor_wavelength])
corrected.plot(legend=False)

corrected.loc[mes_wavelength].to_excel(folder.split("*")[0] + 'experiments.xlsx')
```

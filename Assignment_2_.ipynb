Q. Create a **dataframe** variable **ans_q7** that describes menus in **Pizza Hut** pizza. This dataframe has exactly three columns "food", "size", and "normalized_price".
The **normalized_price** columns shows the price of a pizza divided by the area of it. 

Note: The area of an $X$-inch pizza is given as $X*X*3.14/4$. You may assume that Pizza Hut pizza is of size

*   6 inches (Personal)
*   12 inches (Medium)
*   14 inches (Large)

For non-pizza items, keep this column NaN. Float variables must match sample answer at least by precision 0.001 for each datapoint. Hint: Use apply


````python
# REPLACE THIS LINE WITH YOUR ANSWER TO QUESTION 7
import warnings
warnings.filterwarnings('ignore')

df1 = df[df['DBA']=='Pizza Hut']
def f(x):
    if x == 'Large':
        return 14*14*np.pi/4
    elif x == 'Medium':
        return 12*12*np.pi/4
    elif x == 'Personal':
        return 6*6*np.pi/4
    else:
        return None

df1['normalized_price'] = df1['price']/df1['size'].apply(lambda x: f(x))
df1 = df1[['food','size','normalized_price']]
df1.reset_index()
````

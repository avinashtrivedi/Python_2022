```python
import random
if __name__ == "__main__":
    ran_num = random.randint(1,100)
    i = 1
    while True:
        num = int(input('Guess a number: '))
        if num<ran_num:
            print('Number is low')
        elif num>ran_num:
            print('Number is high')
        else:
            print(f'You guessed it correctly. in {i} attempt.')
            break
        i = i + 1
```

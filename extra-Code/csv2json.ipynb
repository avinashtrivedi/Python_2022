```python
import json
import csv
import pandas as pd

data_files = sorted(glob('data*.csv'))

['data1.csv', 'data2.csv', 'data3.csv']

mergedata = pd.concat(pd.read_csv(datafile).assign(sourcefilename = datafile)
                     for datafile in data_files)

df = mergedata

def get_nested_rec(key, grp):
    rec = {}
    rec['acctid'] = key
    rec['positions'] = []
    rec['transactions'] = []

    for index, row in grp.iterrows():
        rec['positions'].append({
            "possymbol": row["possymbol"],
            "posdescription": row["posdescription"],
            "posquantity": row["posquantity"],
            "posprice" : row["posprice"],
            "positionvalue": row["positionvalue"],
            "posclass": row["posclass"],
            "posallocation": row["posallocation"]
            
            }),
        
    for index, row in grp.iterrows():
        rec['transactions'].append({
            "transdate": row["transdate"],
            "transtype": row["transtype"],
            "transdescription": row["transdescription"],
            "transamount" : row["transamount"],
            
            }) 
       
    return rec
    
records = []
for key, grp in df.groupby(['acctid']):
    rec = get_nested_rec(key, grp)
    records.append(rec)


with open("demoaccts.json", "w") as outfile:
    json.dump(records, outfile, indent = 4)
```

```python
#here I introduce my data, Pride and Prejudice by Jane Austen
def load_txt(file):
    with open(file,encoding='utf8') as f:
        return ' '.join(line.strip() for line in f.readlines())
```

```python
pride_txt = load_txt('pride.txt')
```

```python
pride_raw = pride_txt[1500:29250]
```

```python
pride_tokens = nltk.sent_tokenize(pride_raw)
```

```python
#here I take words rather than sentences.

pride_word_tokens = nltk.word_tokenize(pride_raw)
pride_words = [w.lower() for w in pride_word_tokens if w.isalnum()]
#print(sorted(set(pride_words)))
```

```python
pride_no_stop = [t for t in pride_words if t not in stopwords.words('english')]
#print(pride_no_stop[:20])
```

```python
fd = nltk.FreqDist(pride_no_stop)
print(fd.most_common(110))
```

```python
#choosing ten words
chosen_words_pride = ['pride', 'young', 'nonsense', 'fortune', 'admired', 'daughters', 'man', 'handsome', 'character', 'friend'] 
```

```python
similarity_dic = {}


for word in chosen_words_pride:
    print(word)
    try:
        similar =  model.most_similar(word, topn = 3)
    except Exception:
        similar = 'not in vocab'
    similarity_dic[word] = similar
```

## Task 2

<b> Using the tools introduced in Task 4 of Workshop 4, build a language model for your corpus
and discuss some of the sentences that can be generated from this language model.</b>

```python
import string
from nltk.lm.preprocessing import padded_everygram_pipeline
from nltk.lm import MLE
```

```python
#Part A:Preparing the data
pride_tokens #I had already tokenised the data above.
```

```python
#remove punctuation
pride_tokens_no_punct = [''.join(letter for letter in word if letter not in string.punctuation) for word in pride_tokens]
pride_tokens_no_punct
```

```python
#seperate sentences
pride_separate_sentences = list((mysent.split() for mysent in pride_tokens_no_punct))
pride_separate_sentences
```

```python
#Part B Generating the model

# Training the model using the nltk n-gram pipeline
# Here "3" means 3-gram - you can change this.
# Note that the sentences start <s> and end </s> tags are inserted automatically
train, vocab = padded_everygram_pipeline(3, pride_separate_sentences)
```

```python
# Instantiate the Maximum Likelihood Estimator (MLE) language model(lm) with the highest order n-gram - here "3" for 3-gram
lm = MLE(3)
```

```python
# Fit the model to generate the vocabulary 
lm.fit(train, vocab)
print(lm.vocab)
```

```python
# Check the 121 th sentence
lm.vocab.lookup(pride_separate_sentences[120])
```

```python
# Check which words are in the vocabulary - <UNK> means unknown
lm.vocab.lookup(["coffee", "assistance", "woman", "cantonese"])
```

```python
# Shows the number and order of n-grams
print(lm.counts)
```

```python
# Displays the count of a specific word
lm.counts['pride']
```

```python
lm.counts['house']
```

```python
# Displays the count of a bigram if the model has bigrams
lm.counts[['can']]['you']
```

```python
# Displays the count of a trigram if the model has trigrams
lm.counts[["can","you"]]["come"]
```

```python
# Displays the score of a trigram if the model has trigrams
lm.score("me", ["you","give"])
```

```python
#now generate sentences using the language model
#you can define the length of the sentence and you can include a text_seed to provide a starting word. Note that typically these will produce different sentences from your model.
```

```python
lm.generate(4)
```

```python
lm.generate(5, text_seed=['she'])
```

```python
lm.generate(6, text_seed=['tell'])
```

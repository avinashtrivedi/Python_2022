```python
def get_five(data,i):
    first_element = []
    x = i-5 if i>=5 else 0
    y = i+5 if i+5<len(data) else len(data)-1
    for indx in range(x,y+1):
        first_element.append(data[indx][0])
    return first_element

with open('feelings.txt', encoding='utf-8') as f:
    data = f.readlines()
data = [line.split() for line in data]
```

```python
lst = []
for i in range(len(data)):
    values = data[i]
    if values[2][0].lower()=='d':
        lst.append(get_five(data,i))
```

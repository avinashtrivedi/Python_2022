```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('Dummy Data.xlsx')
df = df.dropna()
```

```python
def generate_corr_table(n):
    roll_corr_df = pd.DataFrame()
    for col in list(df)[2:]:
        corr = []
        date_list = []
        for i in range(len(df)-n+1):
            x = df[list(df)[1]][i:i+n] # apple
            y = df[col][i:i+n] # other columns , changing
            date = df['Date'][i+n-1]
            rolling_corr = np.corrcoef(x,y)[0][1]
            corr.append(rolling_corr)
            date_list.append(date)
        roll_corr_df[col] = corr
        roll_corr_df['Date'] = date_list
    roll_corr_df.set_index('Date', inplace=True)
    roll_corr_df.reset_index(inplace=True)
    return roll_corr_df
```

```python
generate_corr_table(30)
```

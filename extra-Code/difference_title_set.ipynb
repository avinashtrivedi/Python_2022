```python
def text_prepare(text):
    """
    text: a string
        return: modified initial string
    """
    lower = text.lower() # lowercase text
    space_replace = REPLACE_BY_SPACE_RE.sub(" ",lower) #replace REPLACE_BY_SPACE_RE symbols by space in text
    nosymb = BAD_SYMBOLS_RE.sub("",space_replace) # delete symbols which are in BAD_SYMBOLS_RE from text
    text = re.sub(r"\s*\b(?:{})\b".format("|".join(STOPWORDS)), "", nosymb) # delete STOPWORDS
    return re.sub(r" {2,}", " ", text.strip())
```

```python
def euclidean_dist(x, y):
    """
    Function takes two vectors, x and y, and calculates the Euclidean or straight-line distance between them.

    Pro tip:
    This function calculates the distance by iterating through each element of the two vectors, which mirrors
    the standard equation form of the distance. But a faster way to do this is to use numpy like this:
    return np.linalg.norm(x - y)
    """

    # It's always a good idea to check and make sure your vectors/docs have the same number of dimensions/word types.
    assert len(x) == len(y), print('eucliden_dist(x, y): x and y must be have the same number of dimensions!')

    dist = 0.0

    # Pro tip: zip() allows us to iterate through elements from two vectors/lists in the same order.
    for x_i, y_i in zip(x, y):

        # dist += i is a shorter way of writing dist = dist + i. Exponents are done with two asterisks in python.
        dist += (x_i - y_i)**2

    # After summing the square differences, take the square root (math.sqrt()) and you've got the distance.
    return math.sqrt(dist)


def cosine_sim(x, y):
    
    from scipy import spatial
    cosine_sim = 1 - spatial.distance.cosine(x, y)
   
    # It's always a good idea to check and make sure your vectors/docs have the same number of dimensions/word types.
    assert len(x) == len(y), print('cosine_sim(x, y): x and y must be have the same number of dimensions!')

    # The dot product between x and y. Ie, sum( x_i * y_i for x_i, y_i in zip(x, y) )
    dot_product = np.dot(x, y)

    # Recall that to calculate the magnitude of a vector, we square each element, add them all up, and then take the
    # squre root. Ie, magnitude of vector v is square_root of sum( v_i**2 for v_i in v).
    x_magnitude = math.sqrt(np.sum(np.square(x)))
    y_magnitude = math.sqrt(np.sum(np.square(y)))

    # The cosine similarity is the dot product of x and y divided by the product of their magnitudes.
    return dot_product / (x_magnitude * y_magnitude)


def cosine_dist(x, y):
   
    from scipy import spatial
    cos_dist = spatial.distance.cosine(x, y)

    return 1.0 - cosine_sim(x, y)

print()
```

```python
 # Gensim's dictionary object creates a mapping from each word type to an integer. This integer tells us which
    # dimension or element index the word type has been assigned to. The dictionary keys are the integers (or word IDs)
    # and the values are strings (or word types).
gensim_dictionary = gensim.corpora.Dictionary(titles_tokens_list)

    # The length of this dictionary tells us how many word types there are, ie, vocabulary size. Note that this is based
    # on all of the documents we are using.
vocab_size = len(gensim_dictionary)

    # As an example, let's loop through the 100 first word types in the dictionary.
for wid in gensim_dictionary.keys()[:100]:
        print('word ID (dimension) ' + str(wid) + ' -> ' + gensim_dictionary[wid])
print('.\n.\n.')
for wid in gensim_dictionary.keys()[-100:]:
        print('word ID (dimension) + ' + str(wid) + ' -> ' + gensim_dictionary[wid])
print()
```

```python
  # gensim_corpus will be a list of lists, where each element is a list of these (word id, word count) tuples.
gensim_corpus = [gensim_dictionary.doc2bow(titles_tokens) for titles_tokens in titles_tokens_list]

    # We can convert from gensim's memory-efficient representation to a full matrix representation that takes more
    # memory, but is ready to do math. The corpus2dense() function returns a term-by-document matrix by deafult. If we
    # add a ".T" to the end, we get a document-by-term matrix instead.
matrix_corpus = gensim.matutils.corpus2dense(gensim_corpus, vocab_size, num_docs=len(gensim_corpus)).T

print('The matrix has shape ' + str(matrix_corpus.shape) + ', reflecting ' + str(matrix_corpus.shape[0]) +
          ' documents and ' + str(matrix_corpus.shape[1]) + ' word types.\n')

   
```

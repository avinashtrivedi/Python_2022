```python
import matplotlib.pyplot as plt
import numpy as np

# Some example data to display
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
y1 = np.cos(x ** 2)
```

```python
fig, ax = plt.subplots(5,figsize=(10,10))
ax[0].plot(x, y,label='bb')
ax[0].plot(x, y1,label='rr')

ax[0].legend()
ax[1].plot(x + 1, -y)
```

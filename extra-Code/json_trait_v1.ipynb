```python
import json
import os
import pandas as pd
pd.options.display.max_colwidth = 500

# change this file location.
path = r'C:\Users\avitr\OneDrive\Desktop\a3\New folder\json'
```

```python
filenames = os.listdir(path)
df = pd.DataFrame({'fname':filenames})
```

```python
def get_str(fname):
    
    filepath = os.path.join(path,fname)
    with open(filepath) as fp:
        data = json.load(fp)

    s_total = ''
    for i in range(len(data['traits'])):
        key = data['traits'][i]['trait_type']
        val = data['traits'][i]['value']
        s = key + ': ' + val + ', '
        s_total = s_total + s
    return s_total[:-2]
```

```python
df['trait_type_value'] = df['fname'].apply(lambda x: get_str(x))
df['fname'] = df['fname'].apply(lambda x:x.split('.')[0])
```

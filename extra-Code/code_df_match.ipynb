```python
import pandas as pd
```

```python
df1 = pd.read_csv('TDMX_DOCUMENT_S_9501.csv')
df2 = pd.read_excel('TDMX_DOCUMENTS_S_9501_UI.xlsx')
```

```python
df1.head()
```

```python
df2.head()
```

```python
df1_match = set()
df2_match = set()
for index1, row1 in df1.iterrows():
    for index2, row2 in df2.iterrows():
        if row1['tdmx_id'] == row2['objectid'] and row1['revision'] == row2['revision']:
            df1_match.add(index1)
            df2_match.add(index2)
```

```python
df1_non_match = set(df1.index)-df1_match
df2_non_match = set(df2.index)-df2_match
```

```python
df1.iloc[list(df1_match)].to_csv('MATCH_TDMX_DOCUMENT_S_9501.csv')
df1.iloc[list(df1_non_match)].to_csv('NONMATCH_TDMX_DOCUMENT_S_9501.csv')

df2.iloc[list(df2_match)].to_csv('MATCH_TDMX_DOCUMENTS_S_9501_UI.csv')
df2.iloc[list(df2_non_match)].to_csv('NON_MATCH_TDMX_DOCUMENTS_S_9501_UI.csv')
```

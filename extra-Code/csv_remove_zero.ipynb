```python
import pandas as pd
import numpy as np

# pd.set_option('display.max_rows', 2000)
# pd.set_option('display.max_columns', 3000)
# pd.set_option('display.width', 1000)
```

```python
df = pd.read_csv('transposed.csv')
```

```python
df.head()
```

```python
df_list_data = []
col_name = []
for col in list(df):
    try:
        df_col_list = df[col].to_list()
        index = df_col_list.index(1)
        df_list_data.append(df_col_list[index:])
        col_name.append(col)
    except:
        pass   
```

```python
row = max([len(data) for data in df_list_data])
```

```python
for i in range(len(df_list_data)):
    df_list_data[i] = df_list_data[i] + [float('NaN')]*(row - len(df_list_data[i]))
```

```python
df_list_data = np.array(df_list_data).T
```

```python
df_final = pd.DataFrame(df_list_data,columns=col_name)
```

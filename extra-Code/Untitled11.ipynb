```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('clean_covid_data (1).csv')
df = df[['Country','TotalCases','TotalDeaths']]
```

```python
df['TotalDeaths'] = df['TotalDeaths'].apply(lambda x: 0 if x==' ' else x)
df['TotalDeaths'] = df['TotalDeaths'].astype('int')
```

```python
x = np.arange(len(df))
plt.figure(figsize=(35,5))
y1 = df['TotalCases']
y2 = df['TotalDeaths']
# plt.bar(x, y1, color='red',width=0.3,label='TotalCases')
# plt.bar(x, y2, color='green',width=0.3,label='TotalDeaths')

plt.scatter(x, y1, color='red',label='TotalCases')
plt.scatter(x, y2, color='green',label='TotalDeaths')

plt.xticks(x, list(df['Country']))
plt.xticks(rotation=90)
plt.title('Covid19 Cases')
plt.xlabel('Country')
plt.ylabel('Total No. of Cases')
plt.legend()
plt.show()
```

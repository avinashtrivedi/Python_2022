```python
def display_menu():
    print("1. View all\n2. Add item\n3. Delete item\n4. Quit")

def view_all(items):
    print("\nVIEW ALL")
    print("--------------------------------")
    print("NAME\t\tPRICE\tQUANTITY")
    print("--------------------------------")
    tot_items=0
    tot_cost=0
    #traversing through each item in items list
    for item in items:
        price=item["price"]
        quantity=item["quantity"]
        tot_items+=quantity #adding to quantity
        tot_cost+=price*quantity #tot_cost is quantity * price
        #str.ljust(x,"y") is used for formatting the string
        #str will be appended to the left with 12 spaces
        #format(x,".2f") number x will have 2 digits after decimal
        print(item["name"].ljust(12," "),"\t","$"+format(price,".2f"),"\t",quantity,sep="")
    print("\n-- Total Items:",tot_items)
    print("\n-- Total Cost:",format(tot_cost,".2f"))

def add_item(items):
    print("\nADD ITEM")
    name = input("Enter item name: ")
    price = float(input("Enter item price: "))
    quantity = int(input("Enter item quantity: "))
    d={"name":name,"price":price,"quantity":quantity}
    #appending new item to items
    items.append(d)

def get_item_index(name,items):
    for i in range(len(items)):
        #if current items name is same as name
        if items[i]["name"]==name:
            return i
    #when item is not found
    return -1

def remove_item(items):
    print("\nDELETE ITEM")
    #empty list
    if len(items)==0:
        print("NO items to remove.")
    else:
        name=input("Enter item name: ")
        index=get_item_index(name,items)
        #getting name till the item is found
        while(index==-1):
            name=input("Enter item name: ")
            index=get_item_index(name,items)
        print("\nRemoving the following item:")
        print(">> Name:",items[index]["name"])
        print(">> Price:","$"+str(items[index]["price"]))
        print(">> Quantity:",items[index]["quantity"])
        confirm = input("\nConfirm? (y/n)")
        if confirm=="y":
            #pop will remove the item from list
            items.pop(index)
            print("\n-- Item removed!")
        else:
            print("\n-- Item not removed!")

def main():
    print("INVENTORY")
    display_menu()
    op = int(input(">> Enter option: "))
    items=[]
    #loop will continue will 4 is entered
    while(op!=4):
        if op==1:
            view_all(items)
        elif op==2:
            add_item(items)
        elif op==3:
            remove_item(items)
        print("\nINVENTORY")
        display_menu()
        op = int(input(">> Enter option: "))
    #end of program
    print("Good bye!!!")

main()
```

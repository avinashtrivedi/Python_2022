```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel('pc.xlsx',sheet_name='SIX Bitcoin ETPs')
```

```python
col = ['Dates',
 'ABTC TURNOVER',
 'ABTC AVERAGE_BID',
 'BTCEUSD TURNOVER',
 'BTCEUSD AVERAGE_BID',
 'VBTC TURNOVER',
 'VBTC AVERAGE_BID',
 'BTCW TURNOVER',
 'BTCW AVERAGE_BID',
 'BITC  TURNOVER',
 'BITC  AVERAGE_BID',
 'SBTCU TURNOVER',
 'SBTCU AVERAGE_BID',
 'BTIC TURNOVER',
 'BTIC AVERAGE_BID',
 'XBTIUSD TURNOVER',
 'XBTIUSD AVERAGE_BID',
 'FBTC TURNOVER',
 'FBTC AVERAGE_BID']
```

```python
df.columns = col
```

```python
df = df.iloc[2:,:]
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
df
```

```python
df.isna().sum()
```

```python
df['Dates'] = pd.to_datetime(df['Dates'])
```

```python
df.dtypes
```

```python
for col in list(df)[1:]:
    df[col] = df[col].astype('float')
```

```python
df.dtypes
```

```python
df['Year'] = df['Dates'].dt.year 
df['Month'] = df['Dates'].dt.month 
```

```python
dict_mnth = {
    1:'JAN',
    2:'FEB',
    3: 'MARCH',
    4: 'APRIL',
    5: 'MAY',
    6: 'JUNE',
    7: 'JULY',
    8: 'AUG',
    9: 'SEP',
    10: 'OCT',
    11: 'NOV',
    12: 'DEC'
}
```

# Turnover

```python
turnover_col = [name for name in list(df) if 'TURNOVER' in name]
df_new = df.groupby(['Year','Month'])[turnover_col].sum()
df_new = df_new.reset_index()
```

```python
df_new
```

```python
yr_month = []
for _,row in df_new.iterrows():
    
    yr_month.append(str(row['Year']).split('.')[0]+'_'+dict_mnth[row['Month']])
    
df_new['yr_month'] = yr_month
```

```python
fig, ax = plt.subplots(figsize=(10,5))
# turnover_col = [name for name in list(df_new) if 'TURNOVER' in name]
new_name = df_new['yr_month'].iloc[[0,2,4,6,8,10,12]]
for col in turnover_col:
    ax.plot(list(range(len(df_new))),list(df_new[col]),label=col)
ax.set_xticklabels(new_name)
plt.legend()
plt.show()
```

# Average bid

```python
bid_col = [name for name in list(df) if 'AVERAGE_BID' in name]
```

```python
df_bid = df.groupby(['Year','Month'])[bid_col].mean()*100
df_bid = df_bid.reset_index()
```

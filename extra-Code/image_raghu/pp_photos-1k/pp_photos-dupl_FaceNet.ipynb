## Face Extraction and compute embeddings for images using FaceNet

```python
from numpy import load
from numpy import expand_dims
from numpy import asarray
from numpy import savez_compressed
from keras.models import load_model
from keras_facenet import FaceNet

embedder = FaceNet()
```

```python
help(FaceNet())
```

```python
train_images_path = []
trainy =[]
embedding = []
def FaceNetEmbedding(directory,threshold=0.95):
    for subdir in listdir(directory):
        subpath = os.path.join(directory,subdir)
        images = listdir(subpath)
        for image in images:
            image_path = os.path.join(subpath,image)
            trainy.append(image)
            train_images_path.append(image_path)
            embedding.append(embedder.extract(image_path,threshold=threshold)[0]['embedding'])
            print(f'Embedding Created for {subdir}/{image}')
    return np.array(embedding)
```

```python
features_dataset = FaceNetEmbedding(dataset_train_path,0.95)
```

```python
# save arrays to one file in compressed format
savez_compressed('./faces-train_embeddings.npz', features_dataset, trainy, train_images_path)
```

## Finding Duplicate Images

```python
#load dataset
data = load('./faces-train_embeddings.npz')
trainX_features, trainy, train_images_path = data['arr_0'], data['arr_1'], data['arr_2'] #, data['arr_3']
```

```python
def get_duplicate_images_df(feature, matching_cutoff_threshold = 0.8):
    img_dup = pd.DataFrame([],columns=['image_id','Image_label','Image_path','Duplicate_image','Duplicate_path','similarity_score'])
    for id1,feature1 in tqdm(enumerate(zip(feature['arr_0'], feature['arr_1'], feature['arr_2']))):
        for id2,feature2 in enumerate(zip(feature['arr_0'], feature['arr_1'], feature['arr_2'])):
            if id2>id1:
                image_1,label_1,path_1 = feature1[0],feature1[1],feature1[2]
                image_2,label_2,path_2 = feature2[0],feature2[1],feature2[2]

                similarity_score = cosine_similarity(image_1.reshape(1, -1),
                                                     image_2.reshape(1, -1))
                if similarity_score > matching_cutoff_threshold:
                    img_dup.loc[len(img_dup)]= [id1,label_1,path_1,label_2,path_2,similarity_score[0][0]]
    return img_dup
```

```python
df_dup = get_duplicate_images_df(data)
```

# Save images to Folder

```python
image_org = set(df_dup['Image_path'])
image_dup = set(df_dup['Duplicate_path'])
total_image = image_org.union(image_dup)

# generate random folder name to save original and duplicate
foldername = ''.join(random.choice(string.ascii_uppercase) for i in range(5)) + '_OutPut'
newdir = './' + foldername

assert os.path.exists(newdir) == False , 'Folder Already Exist, Please Create new folder'

# create a folder
os.makedirs(newdir)

for image_path in total_image:    
    
    # .split('/')[-1] for without sub folder
    dest_path = os.path.join(newdir,image_path.split('\\')[-1])
    # copy image to dest_path
    _ = copyfile(image_path,dest_path)
print('Output folder:->',newdir)
```

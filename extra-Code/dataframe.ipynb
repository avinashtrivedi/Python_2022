```python
# 1
import string
letters = list(string.ascii_lowercase)
```

```python
# 2
data = [0]*len(letters)
df_letters = pd.DataFrame(data,index=letters,columns=['Letter_Counts'])
df_letters
```

```python
# 3
```

```python
example_transcript = df['transcript'][0].lower()
```

```python
print(example_transcript)
```

```python
# 4
```

```python
for each in example_transcript:
    if each in letters:
        df_letters.loc[each] += 1
```

```python
df_letters
```

```python
# 5
```

```python
plt.figure(figsize=(10,5))
plt.bar(df_letters.index, df_letters['Letter_Counts'],color='blue')
plt.show()
```

```python
df_letters.sort_values('Letter_Counts',ascending=False).head()
```

```python
df_letters['Letter_Counts%'] = df_letters['Letter_Counts'].apply(lambda x: x*100/sum(df_letters['Letter_Counts']))
```

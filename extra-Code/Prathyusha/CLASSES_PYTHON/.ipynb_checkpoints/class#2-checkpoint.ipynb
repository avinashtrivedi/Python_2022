# if - else

```python
while True:
    while True:
        a = input('Enter a integer: ')

        if a.isnumeric():
            a = int(a)
            if a==4:
                print('i am 4')
            elif a==5:
                print('i am 5')
            else:
                print('NA')
            break
        else:
            print('Wrong input')
    break
```

```python
for i in range(10):
    if i==5:
        print('Hello')
        break
else:
    print('Hi')
```

```python
for i in range(6,10):
    if i==5:
        print('Hello')
        break
else:
    print('Hi')
```

```python
for i in range(10):
    if 0:
        print('Hello')
        break
else:
    print('Hi')
```

```python
x = [55,13,66,88]
for i in range(10):
    if i in x:
        break
else:
    print('Hi')
```

# List comprehension

```python
for i in range(10):
    if i%2==0:
        print('even')
    else:
        print('odd')
```

```python
[i for i in range(10) if i%2==0]
```

```python
for i in range(10):
    if i%2==0:
        print(i)
```

```python
['even' if i%2==0 else 'odd' for i in range(10)]
```

```python
[i**2 if i%2==0 else i**3 for i in range(10)]
```

```python
[[1,2,3],[4,5,6],[7,8,9]]
```

```python
[[j for j in range(3)] for i in range(3)]
```

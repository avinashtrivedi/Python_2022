```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import re
import pandas as pd
from nltk import everygrams, word_tokenize
import numpy as np
```

```python
df = pd.read_csv('New MCR Data- 0401 - Sheet2.csv')
```

```python
s = df['Summary'][4]
s = [i.lower() for i in s.split('\n') if i!='']
```

```python
def clean_text(sentences):
    result = []
    for sentence in sentences:
        sentence = re.sub('[^a-zA-Z]', ' ', sentence)
        sentence = re.sub("\s+[a-zA-Z]\s+", ' ', sentence)
        sentence = re.sub('\s+', ' ', sentence)
        sentence = sentence.strip().lower()
        result.append(sentence)
    return result
```

```python
def stop_and_stem(text):
#     ps = PorterStemmer() 
    lemmatizer = WordNetLemmatizer() 

    '''
    Removes stopwords and does stemming
    '''
    stoplist = stopwords.words('english')
#     stoplist.extend(common)
    
    text_stemmed = [lemmatizer.lemmatize(word) for word in word_tokenize(text) if word not in stoplist and len(word) > 3]
    text = ' '.join(text_stemmed)
    return text

# twentyone_data['Purpose'] = twentyone_data['Purpose'].apply(stop_and_stem)
```

```python
df['text'] = df['Summary'].apply(lambda x:get_text(x))
```

```python
df["text_clean"] = clean_text(df["text"].tolist())
df["text_clean"] = df["text_clean"].apply(stop_and_stem)
```

```python
def what(x):
    if not x:
        return False
    
    if 'must' in x or 'must not' in x or 'may' in x or 'should' in x or 'could' in x or 'can' in x or 'cannot' in x:
        return True
    else:
        return False
```

```python
df['What'] = df['text'].apply(lambda x:what(x))
```

```python
def who(x):
    if not x:
        return False
    
    if 'wells fargo' in x or 'account holder' in x or 'bold text' in x or 'body text' in x:
        return True
    else:
        return False
```

```python
df['Who'] = df['text'].apply(lambda x:who(x))
```

```python
def when(x):
    if not x:
        return False
    
    if 'when' in x or 'for' in x or 'in' in x or 'as' in x or 'if' in x or 'at 'in x:
        return True
    else:
        return False
```

```python
df['When'] = df['text'].apply(lambda x:when(x))
```

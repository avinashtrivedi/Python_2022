```python
import xml.etree.ElementTree as ET
import pandas as pd
tree = ET.parse('Q7_Activity_Report(2022-07-14 00_00_51).xml')
root = tree.getroot()
root = list(root)[3]
root = list(root)[0]
row_data = [child for child in root if child.tag[-3:]=='Row']

columns = []
for cell in row_data[0]:
    for data in cell:
        if data.text:
            columns.append(data.text)
            
df = pd.DataFrame([],columns=columns)

for row in row_data[1:]:
    mydata = []
    if len(list(row))>1:
        for cell in row:
            if len(list(cell))!=0:
                for data in cell:
                    mydata.append(data.text)
        df.loc[len(df)] = mydata
df.to_csv('final_data.csv')
```

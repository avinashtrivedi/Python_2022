```python
import random
print('GAME LIBRARY')
print('-------------')
print('LOGIN')
u_word = 'student'
p_word = 'abc123'
user_name = input('Username: ')
password = input('Password: ')
print('\n')
flag = False
if user_name==u_word and password==p_word:
    #print game menu
    flag = True

else:
    print("** Incorrect username/password. Try again.**")
    count = 1
    while True:
        print('LOGIN')
        user_name = input('Username: ')
        password = input('Password: ')
#         print("** Incorrect username/password. Try again.**")

        if user_name == u_word and password == p_word:
            flag = True
            break 
            #continue
        else:
            print("** Incorrect username/password. Try again.**\n")
            count +=1 
            if count==3:
                print("Too many failed login attempts. Goodbye.")
                break

# decide its successful login or failed.

if flag==True:
    while True:
        print('GAMES')
        print('-------------------')
        print('1. Dice Roll')
        print('2. Logout')
        option = input('>> Enter option: ')
        
        if option=='1':
            name = input('Enter your name to play: ')
            
            score_user = 0
            score_computer = 0
            
            for i in range(3):
                _ = input('>> Press enter to roll the dice')
                print(f'** Round #{i+1} **\n')
                roll_user = random.randint(1,6)
                roll_computer = random.randint(1,6)
                
                print(f"{name} rolled: {roll_user}")
                print(f"computer rolled: {roll_computer}\n")
                
                if roll_computer > roll_user:
                    print('Computer Wins!')
                    score_computer = score_computer + 1
                elif roll_computer < roll_user:
                    print(f'{name} Wins')
                    score_user = score_user + 1
                    
                print('\nScore')
                print('------')
                print(f'{name}: {score_user}')
                print(f'computer: {score_computer}\n')
                
            if score_computer > score_user:
                print('Computer Wins')
            elif score_user > score_computer:
                print(f'{name} Wins')
            else:
                print(f'Its a Draw')
            break
        
        elif option=='2':
            print('Goodbye. See you soon')
            break
        else:
            print('Invalid option try again.')
```

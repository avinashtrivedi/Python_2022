```python
import pandas as pd
import warnings
import matplotlib as mpl
import matplotlib.pyplot as plt
from itertools import groupby
import numpy as np 
%matplotlib inline
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
warnings.filterwarnings('ignore')

# https://stackoverflow.com/questions/31845258/pandas-multi-index-plotting
```

```python
df = pd.read_csv('20220516_21Shares AG.csv')
```

```python
df = df[['cpty_cntry','cpty_type','symbol','traded_val_eur','cpty_side']]
```

```python
df['traded_val_eur'] = df.apply(lambda x: x['traded_val_eur'] if x['cpty_side']=='BUY' else -x['traded_val_eur'],axis=1)
```

```python
df.dropna(inplace=True)
```

```python
x = df.groupby(['cpty_cntry','cpty_type','symbol','cpty_side']).sum()
x = x.reset_index(level=3, drop=True)
```

```python
y = df.groupby(['cpty_cntry','cpty_type','cpty_side']).sum()
y = y.reset_index(level=2, drop=True)
z = df.groupby(['cpty_cntry','cpty_type']).sum()
```

```python

def add_line(ax, xpos, ypos):
#     plt.figure(figsize=(10,5))
    line = plt.Line2D([xpos, xpos], [ypos + .1, ypos],
                      transform=ax.transAxes, color='gray')
    line.set_clip_on(False)
    ax.add_line(line)
#     plt.xticks(rotation=70)

def label_len(my_index,level):
    labels = my_index.get_level_values(level)
    return [(k, sum(1 for i in g)) for k,g in groupby(labels)]

def label_group_bar_table(ax, df):
#     plt.figure(figsize=(10,5))
    ypos = -.1
    scale = 1./df.index.size
    for level in range(df.index.nlevels)[::-1]:
        pos = 0
        for label, rpos in label_len(df.index,level):
            lxpos = (pos + .5 * rpos)*scale
            ax.text(lxpos, ypos, label, ha='center', transform=ax.transAxes)
            add_line(ax, pos*scale, ypos)
#             plt.xticks(rotation=90)
            pos += rpos
        add_line(ax, pos*scale , ypos)
        ypos -= .1

def multiplot(x,size=(35,10)):
    ax = x.plot(kind='bar',figsize=size,color='orange')
    #Below 2 lines remove default labels
    ax.set_xticklabels('')
    ax.set_xlabel('')
    # ax.xaxis.set_tick_params(rotation=90)
    ax.ticklabel_format(axis="y", style='plain')
    ax.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
    label_group_bar_table(ax, x)
```

```python
multiplot(x)
```

```python
multiplot(y,(22,5))
```

```python
ax = multiplot(z,(20,5)) # (20,5) is image shape
```

```python
x = gutenberg.sents('austen-emma.txt')
text = ''
for i in x:
    s = ' '.join(i)   
    text = text + ' ' + s
tokens = word_tokenize(text)
token_pos_tag = nltk.pos_tag(tokens)
len([i for i in token_pos_tag if i[1] in ('NN','NNP','NNS')])
```

# 1. What factors contribute to happiness reports?

- Happiness Report 2019

```python
corr = df_happy_2019.drop('Overall rank',axis=1).corr()
plt.figure(figsize=(8,5))
plt.title(f'Happiness Report {2019}')
x = sns.heatmap(corr,annot=True,annot_kws={"size": 16})
x.set_yticklabels(x.get_ymajorticklabels(), fontsize = 16)
x.set_xticklabels(x.get_xmajorticklabels(), fontsize = 16)
plt.show()
df_happy_2019[df_happy_2019.columns[1:]].corr()['Score']
```

bases on this we can say that happiness report mostly depends upon GDP,Social support and Healthy life expectancy

# 2. What is the relationship between happiness reports and GDP increase?

```python
# df_happy_2019[['Score','GDP per capita']].plot()

# 2019
plt.figure(figsize=(10,5))
plt.plot(range(len(df_happy_2019)),df_happy_2019['GDP per capita'],label='GDP')
plt.plot(range(len(df_happy_2019)),df_happy_2019['Score'],label='Score')
plt.xlabel('GDP')
plt.ylabel('Score')
plt.legend()
plt.show()
```

```python
# 2020

plt.figure(figsize=(10,5))
plt.plot(range(len(df_happy_2020)),df_happy_2020['Logged GDP per capita'],label='GDP')
plt.plot(range(len(df_happy_2020)),df_happy_2020['Ladder score'],label='Score')
plt.xlabel('GDP')
plt.ylabel('Score')
plt.legend()
plt.show()
```

```python
# 2021

plt.figure(figsize=(10,5))
plt.plot(range(len(df_happy_2021)),df_happy_2021['Logged GDP per capita'],label='GDP')
plt.plot(range(len(df_happy_2021)),df_happy_2021['Ladder score'],label='Score')
plt.xlabel('GDP')
plt.ylabel('Score')
plt.legend()
plt.show()
```

```python
# 2022

plt.figure(figsize=(10,5))
plt.plot(range(len(df_happy_2022)),df_happy_2022['Explained by: GDP per capita'],label='GDP')
plt.plot(range(len(df_happy_2022)),df_happy_2022['Happiness score'],label='Score')
plt.xlabel('GDP')
plt.ylabel('Score')
plt.legend()
plt.show()
```

```python
#Score
df_happy_2020 = pd.read_csv('2020.csv')#Ladder score
df_happy_2021 = pd.read_csv('2021.csv')#Ladder score
df_happy_2022 = pd.read_csv('2022.csv')#Happiness score
```

```python
df_happy_2020 = df_happy_2020[['Country name','Ladder score']]
# df_happy_2020['Happiness_rank_2020'] = range(1,len(df_happy_2020)+1)
df_happy_2020.columns = ['country','Score_2020']
```

```python
df_happy_2021 = df_happy_2021[['Country name','Ladder score']]
# df_happy_2021['Happiness_rank_2021'] = range(1,len(df_happy_2021)+1)
df_happy_2021.columns = ['country','Score_2021']
```

```python
df_happy_2022 = df_happy_2022[['Country','Happiness score']]
# df_happy_2022['Happiness_rank_2022'] = range(1,len(df_happy_2022)+1)
df_happy_2022.columns = ['country','Score_2022']
```

```python
x = pd.merge(df_happy, df_happy_2020, on='country')
x = pd.merge(x, df_happy_2021, on='country')
x = pd.merge(x, df_happy_2022, on='country')
```

```python
x.head()
```

```python
df_covid.columns = ['country', 'life_expectancy', 'human_development_index','total_deaths']
x = pd.merge(x, df_lpi_all, on='country')
x = pd.merge(x, df_covid, on='country')
```

```python
x.head()
```

```python
x['Score_2022'] = x['Score_2022'].apply(lambda x: float(x.replace(',',''))/1000)
cor = x.corr()
plt.figure(figsize=(6,6))
y = sns.heatmap(cor,annot=True,annot_kws={"size": 10})
y.set_yticklabels(y.get_ymajorticklabels(), fontsize = 10)
y.set_xticklabels(y.get_xmajorticklabels(), fontsize = 10)
plt.show()
```

based on heatmap,happiness score is highly correlared to life expectancy and human development index,but its not correlated with covid cases.

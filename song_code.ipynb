```python
class Song:
    
    def __init__(self,title,year,rating):
        self.__title = title
        self.__year = year
        self.__rating = rating
        
    def get_title(self):
        return self.__title
    
    def get_year(self):
        return self.__year
    
    def get_rating(self):
        return self.__rating
    
    def set_title(self,title):
        self.__title = title
    
    def set_year(self,year):
        self.__year = year
    
    def set_rating(self,rating):
        self.__rating = rating

def view_all(songs):
    print('PLAYLIST')
    print('--------')

    print("{:<4} {:<25} {:<5} {:<10}".format('#','NAME','YEAR','RATING'))

    for i in range(len(songs)):
        print("{:<4} {:<25} {:<5} {:<10}".format(i+1,songs[i].get_title(),songs[i].get_year(),songs[i].get_rating()))
    print('\n')
        
def add_song(songs):
    print('ADD SONG')
    name = input('Name: ')
    year = input('Year: ')
    rating = input('Rating: ')
    
    obj = Song(name,year,rating)
    songs.append(obj)
    print('-- Song added!\n')
    
def remove_song(songs):
    view_all(songs)
    print('REMOVE SONG')
    select = input('Enter Song #: ')
    _ = songs.pop(int(select)-1)
    
    print('-- Song Removed!\n')
    
def update_rating(songs):
    view_all(songs)
    print('UPDATE RATING')
    print('-------------')
    select = input('Enter Song #: ')
    rating = input('Enter rating: ')
    
    songs[int(select)-1].set_rating(rating)
    print('\n')

def main():
    playlist = []
    obj1 = Song('Man On The Moon',1992,'***')
    obj2 = Song('Hey Ya!',2003,'*')
    playlist.append(obj1)
    playlist.append(obj2)
    
    while True:
        print('MENU')
        print("1. View all \n2. Add song \n3. Remove Song \n4. Update rating \n5. Exit")
        choice = input('Enter option: ')
        print('\n')
        if choice == '1':
            view_all(playlist)
        elif choice == '2':
            add_song(playlist)
        elif choice == '3':
            remove_song(playlist)
        elif choice == '4':
            update_rating(playlist)
        elif choice == '5':
            break

if __name__ == "__main__":
    main()
```

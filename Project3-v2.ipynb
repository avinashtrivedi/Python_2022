```python
df1 = pd.DataFrame([],columns=range(100))
for i in tqdm(range(len(df))):
    vec = sum(df['vectors'][i])/len(df['vectors'][1])
    df1.loc[len(df1)] = vec
```

```python
df1['type'] = df['type']
```

```python
df1.to_csv('final_data.csv')
```

```python
df1['type'] = df1['type'].replace(list(set(df1['type'])),range(len(list(set(df1['type'])))))
```

```python
X = df1.drop('type',axis=1)
y = df1['type']
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, recall_score
```

```python
from sklearn.model_selection import train_test_split
X_train_val, X_test, y_train_val, y_test = train_test_split(X, y, 
                                                    test_size = 0.25,
                                                    stratify=y,
                                                    random_state = 42)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val, 
                                                    test_size = 0.25,
                                                    stratify=y_train_val,
                                                    random_state = 42)
```

```python
C = 100
svc_model1=LinearSVC(loss='hinge', C = 0.1,random_state=42)
svc_model1.fit(X_train, y_train)
y_predict = svc_model1.predict(X_test)
```

```python
print(classification_report(y_test,y_predict))
```

```python
cm = confusion_matrix(y_test, y_predict)
df_cm = pd.DataFrame(cm,columns=list(set(df['type'])))
```

```python
plt.figure(figsize=(15,15))
x = sns.heatmap(df_cm.corr(),annot=True,annot_kws={"size": 16})
x.set_yticklabels(x.get_ymajorticklabels(), fontsize = 16)
x.set_xticklabels(x.get_xmajorticklabels(), fontsize = 16)
plt.show()
```

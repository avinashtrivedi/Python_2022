## 3.2. Payout considering 10 years

```python
# drop NaN rows
pret.dropna(inplace=True)
```

# for first 120 rows

```python
df_simulate_first120 = pd.DataFrame([],columns=list(range(10)))
for i in range(10):
    
    # list to store simulation details for individual portfolio
    sim_data = []
    
    balance = 120
    for n,ret in enumerate(pret[i][:120]):
        payout = max(1.001,balance/(120-n))
        simulate = balance*(1+round(ret,2)) - payout
        balance = balance - payout
        
        # store simulation detail for individual portfolio
        sim_data.append(simulate)
        
    # create column for each portfolio
    df_simulate_first120[i] = sim_data
    
# export to excel file
df_simulate_first120.to_excel('first_simulate_excel.xlsx')
```

# for last 120 rows

```python
df_simulate_last120 = pd.DataFrame([],columns=list(range(10)))
for i in range(10):
    
    # list to store simulation details for individual portfolio
    sim_data = []
    
    balance = 120
    for n,ret in enumerate(pret[i][-120:]):
        payout = max(1.001,balance/(120-n))
        simulate = balance*(1+round(ret,2)) - payout
        balance = balance - payout
        
        # store simulation detail for individual portfolio
        sim_data.append(simulate)
        
    # create column for each portfolio
    df_simulate_last120[i] = sim_data
    
# export to excel file
df_simulate_last120.to_excel('last_simulate_excel.xlsx')
```

# 6. Montecarlo

```python
def floatToString(x):
    return str(int(x)) + r"/" + str(int(100 - x))

class MonteCarloSimulator:
    def __init__(self,
                 mu,
                 cov,
                 nSamples,
                 nTimeSteps):
        """
        Main initializer.

        Args:
            mu: monthly mean vector.
            cov: monthly covariance matrix
            nSamples: number of simulations
            nTimeSteps: number of time steps for each simulation.
        """
        self.__nSamples = nSamples
        self.__nTimeSteps = nTimeSteps

        self.__mu = mu #monthly
        self.__cov = cov #monthly

        self.__simulatedTrajectories = None
        self.__portfolios = None
        self.__dfPortfolios = None

        self.__makeSimulatedTrajectories()
        self.__makePortfolios()

    def getSimulatedTrajectories(self):
        return self.__simulatedTrajectories

    def getPortfolios(self):
        return self.__portfolios

    def getDfPortfolios(self):
        return self.__dfPorfolios

    def plotPortfoliosReturnsVolatilities(self):
        """
        Plots Returns vs Volatilities of all simulated portfolios.

        Returns:

        """
        dfMeanStd = self.getDfPortfolios().groupby(["weights", "n_trajectory"]).agg(["mean", "std"])
        dfMeanStd.columns = dfMeanStd.columns.droplevel()
        dfMeanStd = dfMeanStd.reset_index()
        dfMeanStd["equity"] = dfMeanStd.weights.apply(lambda s: int(s.split("/")[1]))
        dfMeanStd.rename(columns={"mean": "return", "std": "volatility"}, inplace=True)
        dfMeanStd.sort_values("equity", inplace=True)
        return sns.scatterplot(data=dfMeanStd,
                               x="volatility",
                               y="return",
                               hue="weights",
                               alpha=0.5)

    def plotPortfoliosVolatilitiesCompositions(self):
        """
        Plots a boxplot of the volatilities, according to portfolio composition.

        Returns:

        """
        data_volatility = self.getDfPortfolios().groupby(["weights", "n_trajectory"])
        data_volatility = data_volatility.std().reset_index()
        data_volatility.rename(columns={"returns": "volatility"}, inplace=True)
        data_volatility["equity"] = data_volatility.weights.apply(lambda s: int(s.split("/")[1]))
        data_volatility.sort_values("equity", inplace=True)
        return sns.boxplot(data=data_volatility, x="weights", y="volatility")

    def __makeSimulatedTrajectories(self):
        """
        Simulates trajectories, monthly steps. Shape = nSamples x nTimeSteps x nDimensions
        """
        dt = 12
        dr = np.random.multivariate_normal(self.__mu * dt, #yearly
                                           self.__cov * dt, #yearly
                                           size=(self.__nSamples, self.__nTimeSteps)
                                           )

        self.__simulatedTrajectories = dr

    def __makeSinglePortfolio(self, fixedProportion):
        """
        Makes a single portfolio according to a specified proportion of fixed income.

        Args:
            fixedProportion:

        Returns:

        """
        equityProportion = 1 - fixedProportion
        return self.__simulatedTrajectories[:, :, 0] * fixedProportion + self.__simulatedTrajectories[:, :,
                                                                         1] * equityProportion

    def __makePortfolios(self):
        """
        Make a dictionary of Monte Carlo Portfolios by Composition.
        """
        self.__portfolios = {floatToString(fixedPercentage): self.__makeSinglePortfolio(fixedPercentage / 100)
                             for fixedPercentage in range(0, 101, 10)}

        dfs = []
        for weights, trajectories in self.__portfolios.items():
            for nTrajectory, trajectory in enumerate(trajectories):
                dfs.append(pd.DataFrame({"returns": trajectory,
                                         "n_trajectory": nTrajectory,
                                         "weights": weights}))

        self.__dfPorfolios = pd.concat(dfs)
```

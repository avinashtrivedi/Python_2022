```python
from sklearn.feature_extraction.text import TfidfVectorizer
from gensim.parsing.preprocessing import preprocess_string
```

```python
' '.join(preprocess_string("12-year-old Regan MacNeil be"))
```

```python
df['description'].isna().sum()
```

```python
df.isna().mean()*100
```

```python
# To calculate mean use imputer class
from sklearn.impute import SimpleImputer
import numpy as np
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer = imputer.fit(df['tmdb_score'].values.reshape(-1, 1))
df['tmdb_score'] = imputer.transform(df['tmdb_score'].values.reshape(-1, 1))
```

```python
df.drop(['age_certification','seasons'],axis=1,inplace=True)
```

```python
df.drop(['id','imdb_id'],axis=1,inplace=True)
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df['description_clean'] = df['description'].apply(lambda x: ' '.join(preprocess_string(x)))
```

```python
df
```

```python
X = df['description_clean']
```

```python
# converting the textual data to numerical data
vectorizer = TfidfVectorizer()
vectorizer.fit(X)
X = vectorizer.transform(X)
X = X.toarray()
```
